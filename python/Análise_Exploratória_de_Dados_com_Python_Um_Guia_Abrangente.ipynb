**Análise Exploratória de Dados com Python: Um Guia Abrangente**

A Análise Exploratória de Dados (EDA) é uma etapa fundamental no processo de análise de dados e ciência de dados. Envolve a investigação inicial dos dados para descobrir padrões, identificar anomalias, testar hipóteses e verificar suposições com a ajuda de estatísticas resumidas e visualizações.⁷ A EDA auxilia na determinação da melhor forma de manipular as fontes de dados para obter as respostas necessárias, facilitando a descoberta de relações entre as variáveis e a identificação de erros óbvios.⁹

Para ilustrar os conceitos discutidos neste relatório, serão fornecidos exemplos de código em Python utilizando as bibliotecas Pandas, Matplotlib e Seaborn.

**Carregando Dados**

O primeiro passo em qualquer processo de análise de dados é carregar os dados em um formato que possa ser facilmente manipulado. A biblioteca Pandas em Python é amplamente utilizada para esta finalidade, fornecendo estruturas de dados poderosas como o DataFrame.

```
import pandas as pd

# Criar um dicionário com dados de exemplo
# Corrigido: Adicionado dados às listas 'Nome' e 'Cidade'
# As listas devem ter o mesmo tamanho
data = {
    'Nome': ['Ana', 'João', 'Maria', 'Pedro'],  # Lista de nomes adicionada
    'Idade': [25, 30, 22, 35],                  # Lista de idades (exemplo)
    'Cidade': ['Rio', 'São Paulo', 'Recife', 'Curitiba'] # Lista de cidades adicionada
}

# Criar um DataFrame a partir do dicionário
df = pd.DataFrame(data)

# Imprimir o DataFrame
print(df)
```

Este exemplo demonstra a criação de um DataFrame do Pandas a partir de um dicionário Python, representando um conjunto de dados simples. Este é frequentemente o ponto de partida para a manipulação e análise de dados em Python.¹⁰

**O Que é EDA**

A Análise Exploratória de Dados (EDA) é uma abordagem para analisar conjuntos de dados com o objetivo de resumir suas principais características, frequentemente utilizando gráficos estatísticos e outros métodos de visualização de dados.⁷ A EDA é uma filosofia, uma atitude sobre how a análise de dados deve ser realizada, em vez de um conjunto fixo de técnicas.¹² O objetivo principal da EDA é maximizar a compreensão sobre um conjunto de dados, descobrir a estrutura subjacente, extrair variáveis importantes, detectar outliers e anomalias, testar suposições subjacentes, desenvolver modelos parcimoniosos e determinar as configurações ideais dos fatores.¹³

A filosofia por trás da EDA, introduzida por John Tukey, enfatiza manter uma mente aberta a possibilidades alternativas.¹² Tukey frequentemente comparava a EDA a um trabalho de detetive, onde o papel do pesquisador é explorar os dados de tantas maneiras quanto possível até que uma "história" plausível dos dados se torne aparente.¹² Em vez de testar teorias pré-estabelecidas, a EDA concentra-se em descobrir o que os dados podem revelar.¹⁵

Para começar a entender um DataFrame carregado, podemos visualizar suas primeiras linhas e obter informações básicas sobre suas colunas usando os seguintes comandos:

```
# Assumindo que 'df' é o DataFrame carregado na seção anterior

# Visualizar as primeiras 5 linhas do DataFrame
print("Primeiras linhas do DataFrame:")
print(df.head())

# Obter um resumo das informações do DataFrame, incluindo tipos de dados e valores não nulos
print("\nInformações do DataFrame:")
df.info()
```

A função `df.head()` exibe as primeiras linhas do DataFrame, fornecendo uma visão rápida dos dados.¹⁶ O método `df.info()` fornece um resumo do DataFrame, incluindo o tipo de dados de cada coluna e o número de valores não nulos, o que é crucial para identificar potenciais dados ausentes.¹⁰

**Visualização**

A visualização desempenha um papel crítico na EDA, pois permite que os analistas de dados "vejam" seus dados e compreendam as variáveis e as relações entre elas.¹⁹ As representações gráficas traduzem números brutos em insights visuais compreensíveis, facilitando a detecção de tendências, outliers ou relações entre variáveis.²¹

Os seguintes exemplos de código demonstram como criar diferentes tipos de gráficos usando as bibliotecas Matplotlib e Seaborn.

**Histograma**

Um histograma representa a distribuição de uma variável numérica para um ou vários grupos. Os valores são divididos em bins, e cada bin é representado como uma barra.²²

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Criar dados numéricos de exemplo
data_histogram = np.random.randn(1000)

# Usando Matplotlib
plt.figure(figsize=(8, 6))
plt.hist(data_histogram, bins=30, edgecolor='black')
plt.title('Histograma com Matplotlib')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()

# Usando Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(data_histogram, bins=30, kde=True)
plt.title('Histograma com Seaborn')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```

Histogramas são ferramentas fundamentais para entender a distribuição, a tendência central e a assimetria de dados numéricos.²⁴

**Gráfico de Dispersão**

Um gráfico de dispersão é um tipo de gráfico que mostra a relação entre duas variáveis numéricas. Cada membro do conjunto de dados é plotado como um ponto cujas coordenadas x-y se relacionam aos seus valores para as duas variáveis.²⁶

```
# Criar dados de exemplo para gráfico de dispersão
data_scatter_x = np.random.rand(100)
data_scatter_y = 2 * data_scatter_x + np.random.randn(100) * 0.5

# Usando Matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(data_scatter_x, data_scatter_y)
plt.title('Gráfico de Dispersão com Matplotlib')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.show()

# Usando Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data_scatter_x, y=data_scatter_y)
plt.title('Gráfico de Dispersão com Seaborn')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.show()
```

Gráficos de dispersão são cruciais para identificar correlações, tendências e potenciais outliers em dados numéricos bivariados.³

**Gráfico de Barras**

Gráficos de barras são comumente usados para visualizar e comparar variáveis categóricas. Eles usam barras para representar a frequência ou contagem de cada categoria, facilitando a identificação das categorias mais prevalentes e suas proporções relativas.³⁰

```
# Criar dados de exemplo para gráfico de barras
# Corrigido: Adicionado dados à lista de categorias
categories = ['A', 'B', 'C', 'D']
counts = [15, 50, 25, 60] # Ajustado para ter 4 elementos

# Usando Matplotlib
plt.figure(figsize=(8, 6))
plt.bar(categories, counts)
plt.title('Gráfico de Barras com Matplotlib')
plt.xlabel('Categorias')
plt.ylabel('Contagem')
plt.show()

# Usando Seaborn (barplot)
plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=counts)
plt.title('Gráfico de Barras com Seaborn (barplot)')
plt.xlabel('Categorias')
plt.ylabel('Contagem')
plt.show()

# Usando Seaborn (countplot)
# Corrigido: Criando um DataFrame mais representativo para countplot
# Countplot conta ocorrências, então replicamos categorias baseado nas contagens (aproximadamente)
category_list = []
for cat, count in zip(categories, counts):
    category_list.extend([cat] * count) # Simula dados brutos para countplot
data_bar_countplot = pd.DataFrame({'Categoria': category_list})

plt.figure(figsize=(8, 6))
# Passando o DataFrame e especificando a coluna x
sns.countplot(data=data_bar_countplot, x='Categoria', order=categories) # order opcional para manter a ordem
plt.title('Gráfico de Barras com Seaborn (countplot)')
plt.xlabel('Categorias')
plt.ylabel('Contagem')
plt.show()
```

Gráficos de barras são essenciais para entender a frequência ou a proporção de diferentes categorias em um conjunto de dados.⁴

**Mapa de Calor**

Mapas de calor são representações gráficas de dados onde os valores são codificados usando cores. Na EDA, eles são frequentemente usados para visualizar matrizes de correlação, mostrando as relações entre variáveis numéricas.³²

```
# Criar dados de exemplo para mapa de calor (matriz de correlação)
# Gerando dados mais estruturados para ter uma correlação mais significativa
np.random.seed(42) # Para reprodutibilidade
data_heatmap = pd.DataFrame(np.random.rand(100, 5), columns=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'])
data_heatmap['Var3'] = data_heatmap['Var1'] * 2 + np.random.normal(0, 0.1, 100)
data_heatmap['Var5'] = -data_heatmap['Var2'] + np.random.normal(0, 0.2, 100)
correlation_matrix = data_heatmap.corr()

# Usando Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f") # fmt para formatar anotações
plt.title('Mapa de Calor com Seaborn')
plt.show()
```

Mapas de calor são valiosos para identificar rapidamente padrões de correlação entre múltiplas variáveis numéricas.³³

**Etapas Envolvidas em EDA (Coleta de Dados)**

As etapas iniciais da EDA envolvem a coleta e o carregamento dos dados. Compreender a origem e o método de coleta de dados é crucial para interpretar os resultados da EDA e identificar potenciais vieses.⁸

O exemplo de código a seguir demonstra como ler um arquivo CSV para um DataFrame do Pandas e verificar o número de linhas e colunas:

```
# Assumindo que temos um arquivo chamado 'data.csv' no mesmo diretório
# Para este exemplo funcionar, crie um arquivo 'data.csv' simples, por exemplo:
# Nome,Idade,Cidade
# Carlos,45,Natal
# Julia,29,Fortaleza
# Marcos,33,Salvador

# Ler o arquivo CSV para um DataFrame
try:
    df_csv = pd.read_csv('data.csv')
    print("DataFrame carregado do arquivo CSV:")
    print(df_csv.head())

    # Verificar o número de linhas e colunas
    num_linhas, num_colunas = df_csv.shape
    print(f"\nNúmero de linhas: {num_linhas}")
    print(f"Número de colunas: {num_colunas}")
except FileNotFoundError:
    print("Erro: O arquivo 'data.csv' não foi encontrado.")
    print("Por favor, crie um arquivo 'data.csv' de exemplo no mesmo diretório para executar este código.")
    # Criando um DataFrame de exemplo caso o arquivo não exista, para códigos seguintes funcionarem
    df_csv = pd.DataFrame({
        'Nome': ['Carlos', 'Julia', 'Marcos'],
        'Idade': [45, 29, 33],
        'Cidade': ['Natal', 'Fortaleza', 'Salvador'],
        'Renda': [5000, 3500, 4200], # Adicionando colunas para exemplos posteriores
        'Valor': [100, 150, 120],
        'Largura': [10, 12, 11],
        'Altura': [5, 6, 5.5],
        'Cor': ['Azul', 'Verde', 'Azul']
    })
    # Adicionando alguns NaNs para o exemplo de tratamento de valores ausentes
    df_csv.loc[1, 'Idade'] = np.nan
    df_csv.loc[0, 'Renda'] = np.nan
    print("\nUsando DataFrame de exemplo criado em código:")
    print(df_csv.head())
    num_linhas, num_colunas = df_csv.shape
    print(f"\nNúmero de linhas: {num_linhas}")
    print(f"Número de colunas: {num_colunas}")
```

A função `pd.read_csv()` lê dados de um arquivo CSV para um DataFrame do Pandas.³⁴ O atributo `df.shape` retorna uma tupla contendo o número de linhas e colunas no DataFrame, fornecendo uma compreensão básica do tamanho do conjunto de dados.³⁶

**Lidando com Valores Ausentes**

Identificar e lidar com valores ausentes é importante na EDA. Valores ausentes podem afetar a precisão das análises e modelos, exigindo técnicas de tratamento adequadas.³⁸

Os exemplos de código a seguir ilustram como verificar a contagem de valores ausentes por coluna e como imputar valores ausentes usando a média e a mediana:

```
# Usando 'df_csv' do passo anterior (seja lido do arquivo ou criado em código)
df_tratamento = df_csv.copy() # Criar cópia para não alterar o original

# Verificar a contagem de valores ausentes por coluna
missing_counts = df_tratamento.isnull().sum()
print("Contagem de valores ausentes por coluna:")
print(missing_counts)

# Verificar se a coluna 'Idade' existe e tem valores ausentes antes de prosseguir
if 'Idade' in df_tratamento.columns and df_tratamento['Idade'].isnull().any():
    # Calcular a média da coluna 'Idade' (ignorando NaNs)
    mean_age = df_tratamento['Idade'].mean()
    print(f"\nMédia de Idade (para preenchimento): {mean_age:.2f}")

    # Imputar os valores ausentes na coluna 'Idade' com a média
    df_tratamento['Idade_preenchida_media'] = df_tratamento['Idade'].fillna(mean_age)

    print("\nDataFrame com valores ausentes de 'Idade' preenchidos com a média:")
    print(df_tratamento[['Idade', 'Idade_preenchida_media']].head())

    # Calcular a mediana da coluna 'Idade' (ignorando NaNs)
    median_age = df_tratamento['Idade'].median()
    print(f"\nMediana de Idade (para preenchimento): {median_age:.2f}")

    # Imputar os valores ausentes na coluna 'Idade' com a mediana
    df_tratamento['Idade_preenchida_mediana'] = df_tratamento['Idade'].fillna(median_age)

    print("\nDataFrame com valores ausentes de 'Idade' preenchidos com a mediana:")
    print(df_tratamento[['Idade', 'Idade_preenchida_mediana']].head())
else:
    print("\nA coluna 'Idade' não existe ou não possui valores ausentes no DataFrame de exemplo.")
```

O método `df.isnull().sum()` retorna a contagem de valores ausentes (NaN) em cada coluna.³⁹ Os métodos `.mean()` e `.median()` calculam a média e a mediana de uma coluna, respectivamente. O método `.fillna()` é usado para substituir os valores ausentes pelos valores calculados.⁴¹

**Lidando com Outliers**

Outliers podem distorcer medidas estatísticas e afetar negativamente o desempenho do modelo. Portanto, é importante identificar e lidar com eles de forma apropriada.⁴⁴

Os exemplos de código a seguir demonstram como calcular os limites do IQR e identificar outliers em colunas numéricas, bem como um exemplo de como aplicar o capping (limitação) usando `np.clip()` em uma coluna específica como 'Renda'.

```
import numpy as np

# Usando 'df_csv' (ou o DataFrame de exemplo criado se o CSV falhou)
# Certificando que temos a coluna 'Renda' para o exemplo
if 'Renda' not in df_csv.columns:
    df_csv['Renda'] = [5000, 3500, 4200, 15000, 2000] # Adicionando com um outlier
    df_csv.loc[0, 'Renda'] = np.nan # Garantir um NaN para testar a robustez
    print("\nAdicionada coluna 'Renda' de exemplo para teste de outliers.")

df_outlier = df_csv.copy()
coluna_numerica = 'Renda' # Coluna para analisar outliers

# Verificar se a coluna existe e é numérica
if coluna_numerica in df_outlier.columns and pd.api.types.is_numeric_dtype(df_outlier[coluna_numerica]):
    # Remover NaNs temporariamente para cálculo de IQR
    dados_validos = df_outlier[coluna_numerica].dropna()

    if not dados_validos.empty:
        # Calcular o primeiro quartil (Q1) e o terceiro quartil (Q3)
        Q1 = dados_validos.quantile(0.25)
        Q3 = dados_validos.quantile(0.75)

        # Calcular o IQR
        IQR = Q3 - Q1
        print(f"\nAnálise de Outliers para '{coluna_numerica}':")
        print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")

        # Definir os limites para identificar outliers
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        print(f"Limite Inferior (outlier): {limite_inferior}")
        print(f"Limite Superior (outlier): {limite_superior}")

        # Identificar outliers na coluna original (mantendo NaNs como não-outliers aqui)
        outliers_mask = (df_outlier[coluna_numerica] < limite_inferior) | (df_outlier[coluna_numerica] > limite_superior)
        outliers = df_outlier[outliers_mask]
        print(f"\nOutliers identificados em '{coluna_numerica}' usando IQR:")
        if not outliers.empty:
            print(outliers[[coluna_numerica]])
        else:
            print("Nenhum outlier encontrado.")

        # Aplicar capping (limitação) aos outliers na coluna 'Renda'
        # np.clip não lida bem com NaNs diretamente, então aplicamos onde não é NaN
        df_outlier[f'{coluna_numerica}_capped'] = df_outlier[coluna_numerica].copy()
        df_outlier[f'{coluna_numerica}_capped'] = df_outlier[f'{coluna_numerica}_capped'].clip(lower=limite_inferior, upper=limite_superior)

        print(f"\nDataFrame com outliers de '{coluna_numerica}' com capping:")
        print(df_outlier[[coluna_numerica, f'{coluna_numerica}_capped']].head())
    else:
         print(f"\nA coluna '{coluna_numerica}' não contém dados válidos para cálculo de IQR.")

else:
    print(f"\nA coluna '{coluna_numerica}' não existe ou não é numérica no DataFrame.")
```

O método `.quantile()` calcula os percentis dos dados.⁴⁵ O IQR é a diferença entre o 75º e o 25º percentis. Os limites para outliers são tipicamente definidos como 1,5 vezes o IQR abaixo do Q1 e acima do Q3. A função `np.clip()` (ou o método `.clip()` do Pandas) limita os valores de um array ou Série a um intervalo especificado.⁴⁸

**Engenharia de Features**

A engenharia de features é o processo de usar conhecimento de domínio para extrair features (características, propriedades, atributos) de dados brutos.⁵¹ A criação de novas features a partir das existentes pode revelar padrões ocultos e melhorar o poder preditivo dos modelos.⁵²

Os exemplos de código a seguir show como criar uma nova coluna combinando duas existentes ('Largura' e 'Altura' para criar 'Area') e como aplicar a codificação one-hot usando `pd.get_dummies()` em uma coluna categórica ('Cor').

```
# Usando 'df_csv' (ou o DataFrame de exemplo)
df_eng = df_csv.copy()

# Verificar se as colunas 'Largura' e 'Altura' existem
if 'Largura' in df_eng.columns and 'Altura' in df_eng.columns:
    # Criar uma nova coluna 'Area' combinando 'Largura' e 'Altura'
    df_eng['Area'] = df_eng['Largura'] * df_eng['Altura']

    print("DataFrame com a nova coluna 'Area':")
    print(df_eng[['Largura', 'Altura', 'Area']].head())
else:
    print("\nColunas 'Largura' ou 'Altura' não encontradas para criar 'Area'.")

# Verificar se a coluna 'Cor' existe
if 'Cor' in df_eng.columns:
    # Aplicar one-hot encoding na coluna 'Cor'
    # prefix='Cor' adiciona 'Cor_' ao nome das novas colunas
    df_encoded = pd.get_dummies(df_eng, columns=['Cor'], prefix='Cor', dummy_na=False) # dummy_na=False para não criar coluna para NaN

    print("\nDataFrame após one-hot encoding da coluna 'Cor':")
    # Mostrando colunas relevantes para a mudança
    cols_to_show = [col for col in df_encoded.columns if col.startswith('Cor_') or col == 'Cor']
    # Se 'Cor' foi removida por get_dummies (comportamento padrão), adicionamos outras para contexto
    if 'Cor' not in df_encoded.columns:
        cols_to_show = ['Nome'] + [col for col in df_encoded.columns if col.startswith('Cor_')]

    print(df_encoded[cols_to_show].head())
else:
    print("\nColuna 'Cor' não encontrada para aplicar One-Hot Encoding.")
```

Novas colunas podem ser criadas usando operações aritméticas em colunas existentes.⁵³ A função `pd.get_dummies()` converte variáveis categóricas em variáveis dummy/indicadoras.⁵⁶

**Escalonamento e Normalização de Dados**

O escalonamento e a normalização de dados são técnicas usadas para padronizar o intervalo de variáveis independentes ou features de dados.⁵⁸ Isso garante que todas as features contribuam igualmente para o processo de aprendizado do modelo e pode melhorar o desempenho de certos algoritmos.⁵⁹

Os exemplos de código a seguir fornecem exemplos de como aplicar `StandardScaler` e `MinMaxScaler` do scikit-learn a uma coluna numérica do DataFrame (usando a coluna 'Valor' como exemplo).

```
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

# Usando 'df_csv' (ou o DataFrame de exemplo)
df_scale = df_csv.copy()
coluna_scale = 'Valor'

# Verificar se a coluna existe e é numérica
if coluna_scale in df_scale.columns and pd.api.types.is_numeric_dtype(df_scale[coluna_scale]):
    # Lidar com NaNs antes de escalar: preencher ou remover
    # Aqui, vamos preencher com a média para demonstração
    if df_scale[coluna_scale].isnull().any():
        media_valor = df_scale[coluna_scale].mean()
        df_scale[coluna_scale].fillna(media_valor, inplace=True)
        print(f"\nValores ausentes em '{coluna_scale}' preenchidos com a média ({media_valor:.2f}) antes de escalar.")

    # Inicializar o StandardScaler
    scaler_standard = StandardScaler()
    # Ajustar e transformar a coluna 'Valor'
    # O scaler espera um array 2D, por isso usamos [['Valor']]
    df_scale['Valor_scaled_standard'] = scaler_standard.fit_transform(df_scale[[coluna_scale]])

    print("\nDataFrame com a coluna 'Valor' escalonada usando StandardScaler:")
    print(df_scale[[coluna_scale, 'Valor_scaled_standard']].head())

    # Inicializar o MinMaxScaler
    scaler_minmax = MinMaxScaler()
    # Ajustar e transformar a coluna 'Valor'
    df_scale['Valor_scaled_minmax'] = scaler_minmax.fit_transform(df_scale[[coluna_scale]])

    print("\nDataFrame com a coluna 'Valor' escalonada usando MinMaxScaler:")
    print(df_scale[[coluna_scale, 'Valor_scaled_minmax']].head())

else:
    print(f"\nA coluna '{coluna_scale}' não existe ou não é numérica no DataFrame para escalonamento.")
```

O `StandardScaler` remove a média e dimensiona os dados para variância unitária.⁶⁰ O `MinMaxScaler` dimensiona os dados para um intervalo específico, geralmente entre 0 e 1.⁶²

As bibliotecas Python Pandas, Matplotlib e Seaborn fornecem ferramentas poderosas para realizar tarefas de EDA de forma eficiente. Os exemplos de código fornecidos neste relatório ilustram os conceitos e técnicas fundamentais envolvidos na EDA, desde o carregamento de dados e a inspeção inicial até a visualização, o tratamento de valores ausentes e outliers, a engenharia de features e o escalonamento de dados. A exploração contínua e a experimentação com diferentes conjuntos de dados e técnicas levarão a uma compreensão mais profunda dos dados e a insights mais eficazes.
