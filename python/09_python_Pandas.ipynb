**Estudo Detalhado da Biblioteca Pandas do Python para An√°lise de Dados**

**1. Introdu√ß√£o ao Pandas**

**1.1. O que √© Pandas?**

Pandas √© uma biblioteca de software fundamental para a linguagem de programa√ß√£o Python, projetada especificamente para **manipula√ß√£o e an√°lise de dados**. Lan√ßada como c√≥digo aberto sob a licen√ßa BSD ‚Å∑, ela se estabeleceu como uma ferramenta essencial no ecossistema de ci√™ncia de dados em Python. O nome "Pandas" deriva do termo "Panel Data" (Dados em Painel), um conceito econom√©trico para conjuntos de dados multidimensionais, e tamb√©m de "Python Data Analysis".¬≥, ‚Å∏

A biblioteca se destaca por oferecer estruturas de dados de alto desempenho e f√°ceis de usar, juntamente com um conjunto rico de ferramentas para an√°lise de dados. Seu foco principal reside em trabalhar com dados estruturados, que podem ser tabulares (como planilhas ou tabelas SQL), multidimensionais ou heterog√™neos.‚Å∏ Pandas visa ser a ferramenta de an√°lise e manipula√ß√£o de dados de c√≥digo aberto mais poderosa e flex√≠vel dispon√≠vel.‚Åπ

As caracter√≠sticas chave do Pandas incluem sua rapidez, poder de processamento, flexibilidade e facilidade de uso.‚Å∏ Ele fornece duas estruturas de dados prim√°rias:

* **Series:** Um array unidimensional rotulado, capaz de armazenar dados de qualquer tipo (inteiros, strings, floats, objetos Python, etc.). Possui um **√≠ndice** que rotula cada elemento.‚Å∏
* **DataFrame:** Uma estrutura de dados tabular bidimensional, com eixos rotulados (linhas e colunas). Colunas podem ter tipos diferentes. Visualiz√°vel como uma planilha ou tabela SQL.‚Å∏

**1.2. Import√¢ncia na An√°lise de Dados e Ci√™ncia de Dados**

Pandas √© frequentemente considerada a biblioteca *mais crucial* para an√°lise de dados e ci√™ncia de dados em Python.‚Åπ Sua relev√¢ncia deriva de:

* **Manipula√ß√£o Eficiente de Dados Estruturados:** Facilita enormemente o trabalho com dados tabulares.‚Å∏
* **Conjunto Abrangente de Ferramentas:** Cobre todo o ciclo de an√°lise: leitura/escrita, limpeza, transforma√ß√£o, fus√£o, tratamento de dados ausentes, sele√ß√£o, filtragem, agrega√ß√£o, remodela√ß√£o e visualiza√ß√£o.‚Å∏
* **Integra√ß√£o com o Ecossistema Python:** Integra-se perfeitamente com NumPy (sua base), Matplotlib, Seaborn (visualiza√ß√£o) e Scikit-learn (Machine Learning).‚Å∏
* **An√°lise Explorat√≥ria de Dados (EDA):** Essencial para calcular estat√≠sticas, identificar padr√µes, visualizar distribui√ß√µes e entender os dados.‚Å∏
* **Prepara√ß√£o para Machine Learning:** Indispens√°vel para engenharia de features, codifica√ß√£o de vari√°veis, normaliza√ß√£o, etc.‚Åπ

Pandas atua como a ponte entre dados brutos e an√°lises avan√ßadas ou modelos de ML, simplificando tarefas complexas.‚Åπ

**1.3. Estruturas de Dados Fundamentais (Vis√£o Geral)**

**3. Estruturas de Dados Pandas**

**3.1. Series**

Objeto 1D tipo array com um √≠ndice (`index`). Armazena dados homog√™neos (mas pode cair para `dtype=object`). Tamanho imut√°vel, valores mut√°veis.‚Å∏, ‚Åπ, ¬π‚Å∞

* **Cria√ß√£o (`pd.Series()`):**
    * A partir de lista/array NumPy (√≠ndice padr√£o ou customizado).¬π‚Å∞
    * A partir de dicion√°rio (chaves viram √≠ndice).¬π‚Å∞
    * A partir de escalar (requer √≠ndice).¬≤‚Åµ

````
print("\n--- 3.1 Series ---")

# --------------------------------------
# 1Ô∏è‚É£ Criando uma Series a partir de uma lista, com √≠ndice customizado
# --------------------------------------
dados_lista = [10, 20, 30]  # Lista de valores que v√£o compor os dados da Series
indices = ['a', 'b', 'c']   # √çndices personalizados para cada valor da lista

# pd.Series(data, index, name) -> Cria uma Series com os dados e √≠ndices definidos
# - 'data' pode ser lista, array, dicion√°rio, escalar...
# - 'index' √© opcional: se n√£o for passado, ser√° um RangeIndex padr√£o (0, 1, 2, ...)
# - 'name' √© o nome da Series, √∫til para identifica√ß√£o ou exporta√ß√£o
s_lista = pd.Series(data=dados_lista, index=indices, name='MinhaSerie')

print("Series de lista com √≠ndice e nome:")
print(s_lista)

# Resultado esperado:
# a    10
# b    20
# c    30
# Name: MinhaSerie, dtype: int64

# --------------------------------------
# 2Ô∏è‚É£ Criando uma Series a partir de um dicion√°rio
# --------------------------------------
dados_dict = {'x': 100, 'y': 200, 'z': 300}
# Quando o argumento 'data' √© um dicion√°rio:
# - As chaves se tornam os √≠ndices
# - Os valores se tornam os dados da Series
s_dict = pd.Series(dados_dict)

print("\nSeries de dicion√°rio:")
print(s_dict)

# Resultado esperado:
# x    100
# y    200
# z    300
# dtype: int64

# --------------------------------------
# 3Ô∏è‚É£ Criando uma Series a partir de um escalar
# --------------------------------------
# Um escalar √© um √∫nico valor (ex: 5)
# Quando passamos um escalar como 'data' e uma lista como 'index':
# - O valor escalar √© repetido automaticamente para cada √≠ndice
s_esc = pd.Series(5, index=['p', 'q', 'r'])

print("\nSeries de escalar:")
print(s_esc)

# Resultado esperado:
# p    5
# q    5
# r    5
# dtype: int64
````

````
import pandas as pd

# Vamos reutilizar a Series de exemplo
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'], name='MinhaSerie')

# üîç .values ‚Üí Retorna os dados como um array NumPy
print("s.values ‚Üí", s.values)
# Resultado: array([10, 20, 30])
# Obs: √∫til para opera√ß√µes vetorizadas com NumPy

# üîç .index ‚Üí Retorna o objeto Index (os r√≥tulos/√≠ndices)
print("s.index ‚Üí", s.index)
# Resultado: Index(['a', 'b', 'c'], dtype='object')
# Obs: isso √© um objeto especial do Pandas, com m√©todos pr√≥prios (ex: .tolist())

# üîç .dtype ‚Üí Retorna o tipo dos dados armazenados
print("s.dtype ‚Üí", s.dtype)
# Resultado: int64
# Obs: pode ser int, float, object (texto), category, datetime, etc.

# üîç .name ‚Üí Retorna (ou define) o nome da Series
print("s.name ‚Üí", s.name)
# Resultado: MinhaSerie
# Obs: √∫til ao exportar ou ao usar como coluna em DataFrame

# üîç .size ‚Üí N√∫mero total de elementos
print("s.size ‚Üí", s.size)
# Resultado: 3
# Obs: igual a len(s), ou seja, quantidade de elementos

# üîç .shape ‚Üí Formato (tupla), igual ao array NumPy
print("s.shape ‚Üí", s.shape)
# Resultado: (3,)
# Obs: o n√∫mero de linhas. Como Series √© 1D, vem como tupla de um valor
````

üß† Dica R√°pida:
Esses atributos n√£o s√£o m√©todos (n√£o t√™m `()`) ‚Äî eles s√£o como "etiquetas" de metadados embutidas no objeto `Series`. S√£o muito √∫teis pra diagn√≥sticos e para entender rapidamente o que tem ali dentro sem precisar printar tudo.

````
import pandas as pd

print("\nAcesso e Opera√ß√µes em Series:")

# Series de exemplo
s_lista = pd.Series(data=[10, 20, 30], index=['a', 'b', 'c'], name='MinhaSerie')

# üîç Acesso por r√≥tulo
print(f"Acesso por r√≥tulo 'b': {s_lista['b']}")
# ‚Üí Retorna o valor associado ao √≠ndice 'b' (20).
# ‚Üí Isso √© o modo padr√£o e intuitivo de acessar elementos em uma Series.
# ‚Üí √â como acessar um valor em um dicion√°rio Python.

# üîç Acesso por posi√ß√£o (index inteiro)
print(f"Acesso por posi√ß√£o 1: {s_lista.iloc[1]}")
# ‚Üí Usa o m√©todo `.iloc[]` para acessar pela posi√ß√£o (√≠ndice num√©rico zero-based).
# ‚Üí Aqui, posi√ß√£o 1 corresponde ao segundo elemento (√≠ndice 'b').

# üîç Slicing por r√≥tulo
print(f"Slice por r√≥tulo 'a':'b':\n{s_lista['a':'b']}")
# ‚Üí Retorna os elementos de 'a' at√© 'b' (inclusive 'b'!).
# ‚Üí Importante: no slicing por r√≥tulo, o limite superior √© INCLUSIVO.

# üîç Slicing por posi√ß√£o
print(f"Slice por posi√ß√£o 0:2:\n{s_lista.iloc[0:2]}")
# ‚Üí Retorna os elementos da posi√ß√£o 0 at√© 1 (exclui a posi√ß√£o 2).
# ‚Üí Como listas Python, o limite superior √© EXCLUSIVO.

# üîç Indexa√ß√£o Booleana
print(f"Filtro (valor > 15):\n{s_lista[s_lista > 15]}")
# ‚Üí Retorna apenas os valores maiores que 15.
# ‚Üí Isso √© super √∫til em filtros, an√°lises, e prepara√ß√£o de dados.

# üîç Opera√ß√£o vetorizada
print(f"Opera√ß√£o vetorizada (s_lista * 2):\n{s_lista * 2}")
# ‚Üí Multiplica cada valor da Series por 2.
# ‚Üí Opera√ß√µes vetorizadas s√£o eficientes, r√°pidas e seguras.
# ‚Üí Pandas alinha automaticamente os √≠ndices ao fazer opera√ß√µes entre Series.

# üîÑ Convers√£o de tipo com .astype()
s_float = s_lista.astype(float)
print(f"\nSeries convertida para float:\n{s_float}")
# ‚Üí Converte os valores da Series de int para float (10 ‚Üí 10.0).
# ‚Üí .astype() √© muito usado para otimiza√ß√£o, compatibilidade ou opera√ß√µes espec√≠ficas.
# ‚Üí Pode converter para qualquer tipo compat√≠vel: int, float, str, category, etc.
````

````
import pandas as pd
import numpy as np

print("\n--- 3.2 DataFrame ---")

# üß± 1. Criando um DataFrame a partir de um dicion√°rio de listas:
dados_pessoas = {
    'Nome': ['Alice', 'Bob', 'Charlie'],  # Lista de nomes (coluna 1)
    'Idade': [25, 30, 22],                # Lista de idades (coluna 2)
    'Cidade': ['A', 'B', 'A']             # Lista de cidades (coluna 3)
}
df_pessoas = pd.DataFrame(dados_pessoas)
print("DataFrame de dict de listas:")
print(df_pessoas)

# ‚úÖ Esse m√©todo √© o mais direto: as **chaves do dicion√°rio viram colunas** e os **valores (listas)** s√£o os dados.
# ‚úÖ As listas precisam ter o **mesmo comprimento**, sen√£o o pandas vai reclamar.

# üß± 2. Criando um DataFrame a partir de uma lista de dicion√°rios:
dados_registros = [
    {'id': 1, 'valor': 100, 'status': 'ok'},
    {'id': 2, 'valor': 150},  # 'status' ausente aqui ‚Äî pandas preenche com NaN (Not a Number)
    {'id': 3, 'valor': 120, 'status': 'pendente'}
]
df_registros = pd.DataFrame(dados_registros)
print("\nDataFrame de lista de dicts:")
print(df_registros)

# ‚úÖ Cada dicion√°rio representa uma **linha** no DataFrame.
# ‚úÖ As **chaves se tornam os nomes das colunas**.
# ‚úÖ Se alguma chave estiver ausente em um dicion√°rio, pandas preenche com NaN (valor ausente/nulo).

# üß± 3. Criando um DataFrame a partir de um array NumPy 2D:
dados_np = np.random.randn(3, 4)  # Gera uma matriz 3x4 com valores aleat√≥rios da distribui√ß√£o normal padr√£o
indice_linhas = pd.date_range('2024-01-01', periods=3, freq='D')  # √çndices (linhas) como datas sequenciais
nomes_colunas = ['Sensor1', 'Sensor2', 'Sensor3', 'Sensor4']       # Nomes para as colunas

df_sensores = pd.DataFrame(dados_np, index=indice_linhas, columns=nomes_colunas)
print("\nDataFrame de array NumPy:")
print(df_sensores.round(2))  # Arredondando para 2 casas decimais s√≥ pra ficar mais bonito

# ‚úÖ Esse m√©todo √© √∫til quando voc√™ j√° tem os dados em NumPy e quer organizar em tabela.
# ‚úÖ Voc√™ pode passar `index` e `columns` para deixar a tabela leg√≠vel e informativa.

# --------------------------------------
# üìå Recapitulando os modos de cria√ß√£o:
# pd.DataFrame(dicion√°rio de listas)     -> Cada chave vira uma coluna
# pd.DataFrame(lista de dicion√°rios)     -> Cada dict √© uma linha, colunas s√£o unificadas
# pd.DataFrame(array 2D, columns, index) -> Constr√≥i a tabela com base em dados j√° existentes (como NumPy)

# --------------------------------------
# üß† Curiosidade T√©cnica:
# DataFrame = conjunto de Series alinhadas pelo mesmo √≠ndice
# Ou seja, cada coluna √© uma `Series`, e o DataFrame √© um "dicion√°rio" de Series.
````

````
# Vamos explorar os principais atributos de um DataFrame ‚Äî aqui usamos `df_sensores`
# Esse DF foi criado com dados aleat√≥rios de sensores, datas como √≠ndice e 4 colunas num√©ricas.

print(f"\nAtributos de df_sensores:")

# üîπ .shape: mostra a forma da tabela -> (n√∫mero de linhas, n√∫mero de colunas)
print(f"  Shape: {df_sensores.shape}")  # Exemplo: (3, 4)

# üîπ .dtypes: tipo de dado de cada coluna
print(f"  Dtypes:\n{df_sensores.dtypes}")  # Mostra se √© float, int, object, etc.

# üîπ .columns: retorna um objeto Index com os nomes das colunas
print(f"  Columns: {df_sensores.columns}")  # Ex: Index(['Sensor1', 'Sensor2', ...])

# üîπ .index: retorna o √≠ndice (linhas), que neste caso s√£o datas
print(f"  Index: {df_sensores.index}")  # Ex: DatetimeIndex([...], freq='D')

# üîπ .info(): fornece um resumo geral do DataFrame
# Inclui: total de entradas (linhas), colunas, dtypes, quantos n√£o-nulos em cada coluna e uso de mem√≥ria
print("\n  Info:")
df_sensores.info()

# üîπ .describe(): estat√≠sticas descritivas s√≥ para colunas num√©ricas
# Inclui: count, mean, std, min, 25%, 50%, 75%, max
print("\n  Describe:")
print(df_sensores.describe().round(2))  # Arredondado pra facilitar a leitura

# üîπ .values: retorna os dados como array NumPy 2D
# ‚ö†Ô∏è √ötil pra interoperar com bibliotecas que usam NumPy, mas perde r√≥tulos de linha/coluna
print(f"\n  Valores (.values):\n{df_sensores.values.round(2)}")  # Os dados crus como matriz

# üîπ .T: transposi√ß√£o ‚Äî troca linhas por colunas e vice-versa
# Pode ser √∫til em an√°lise explorat√≥ria e visualiza√ß√£o
print(f"\n  Transposto (.T):\n{df_sensores.T.round(2)}")
````

````
# -*- coding: utf-8 -*-
"""
Este bloco de c√≥digo demonstra e explica em detalhes como usar a fun√ß√£o
pd.read_csv do Pandas para ler dados de arquivos CSV (ou similares),
lidando com diversas complexidades comuns encontradas nesses arquivos.
Inclui explica√ß√µes detalhadas sobre a sintaxe e os par√¢metros utilizados.
"""

# Importa a classe StringIO do m√≥dulo io.
# StringIO permite tratar uma string em mem√≥ria como se fosse um arquivo,
# o que √© √∫til para testar ou quando os dados CSV est√£o em uma vari√°vel string.
from io import StringIO

# Importa a biblioteca Pandas. Assume-se que j√° foi importada,
# mas √© uma boa pr√°tica garantir. 'pd' √© o alias padr√£o.
import pandas as pd

# ==============================================================================
# 4. Leitura e Escrita de Dados (Contexto Fornecido)
# ==============================================================================

# Coment√°rios iniciais fornecidos pelo usu√°rio (mantidos para contexto):
# *(Tabela de Resumo dos Atributos Fundamentais omitida para brevidade, j√° presente na vers√£o anterior)*
#
# **4. Leitura e Escrita de Dados**
#
# Pandas IO Tools: Fun√ß√µes pd.read_* e m√©todos df.to_* para diversos formatos.¬≥‚Å∂
#
# **4.1. Lendo Dados**
#
# * **pd.read_csv()**: Para CSV e arquivos delimitados.¬≥‚Å∂
#     * Par√¢metros detalhados: filepath_or_buffer, sep, header, names, index_col,
#       usecols, dtype, encoding, skiprows, nrows, na_values, parse_dates,
#       skipfooter, comment, thousands, decimal, chunksize (para ler em peda√ßos).‚Å∂, ¬≥‚Å∑, ¬≥‚Å∏

print("\n--- 4.1 Lendo Dados com pd.read_csv ---")

# ==============================================================================
# Simula√ß√£o de Dados CSV Complexos
# ==============================================================================

# Cria uma string multi-linhas que simula o conte√∫do de um arquivo CSV.
# Este exemplo inclui v√°rias caracter√≠sticas que exigem par√¢metros espec√≠ficos
# na fun√ß√£o pd.read_csv para serem lidos corretamente:
# 1. Linha de coment√°rio inicial (#).
# 2. Separador de colunas n√£o padr√£o (';').
# 3. Cabe√ßalho na segunda linha (ap√≥s o coment√°rio).
# 4. N√∫meros formatados com separador de milhar ('.') e decimal (',').
# 5. Um valor ausente representado por um caractere espec√≠fico ('?').
# 6. Uma linha de rodap√© no final que deve ser ignorada.
# 7. Datas em formato de string que precisam ser convertidas.
csv_complexo = """# Coment√°rio a ser ignorado
ID;Data;Valor;Cidade
1;2024-01-10;1.250,50;Rio de Janeiro
2;2024-01-11;850,00;S√£o Paulo
3;2024-01-12;?;Salvador
Rodap√© a ser ignorado"""
# Nota: A quebra de linha ap√≥s a √∫ltima linha de dados e antes do rodap√© √© importante
# para que 'skipfooter' funcione como esperado na string simulada.

# ==============================================================================
# Leitura do CSV Simulado com Par√¢metros Detalhados
# ==============================================================================

# Fun√ß√£o: pd.read_csv(filepath_or_buffer, sep=',', delimiter=None, header='infer', names=None, index_col=None, usecols=None, dtype=None, engine=None, converters=None, true_values=None, false_values=None, skipinitialspace=False, skiprows=None, skipfooter=0, nrows=None, na_values=None, keep_default_na=True, na_filter=True, verbose=False, skip_blank_lines=True, parse_dates=False, infer_datetime_format=False, keep_date_col=False, date_parser=None, dayfirst=False, cache_dates=True, iterator=False, chunksize=None, compression='infer', thousands=None, decimal='.', lineterminator=None, quotechar='"', quoting=0, doublequote=True, escapechar=None, comment=None, encoding=None, dialect=None, error_bad_lines=None, warn_bad_lines=None, on_bad_lines=None, delim_whitespace=False, low_memory=True, memory_map=False, float_precision=None, storage_options=None)
# Objetivo Principal: Ler um arquivo CSV (ou similar) e carreg√°-lo em um DataFrame do Pandas.
# Retorna: Um DataFrame do Pandas (ou um iterador se chunksize for especificado).

try:
    # Tentativa de ler os dados usando pd.read_csv com par√¢metros ajustados
    df_csv_complexo = pd.read_csv(
        # --- Par√¢metro 1: filepath_or_buffer ---
        # Fonte dos dados. Pode ser um caminho de arquivo (string), URL (string)
        # ou um objeto 'file-like' (com um m√©todo read()).
        # Aqui usamos StringIO para tratar nossa string 'csv_complexo' como um arquivo.
        StringIO(csv_complexo),

        # --- Par√¢metro 2: sep (ou delimiter) ---
        # O caractere (ou express√£o regular) que separa as colunas no arquivo.
        # Valor: ';' -> Define o ponto e v√≠rgula como separador. O padr√£o √© ','.
        sep=';',

        # --- Par√¢metro 3: skiprows ---
        # N√∫mero de linhas a pular do *in√≠cio* do arquivo antes de come√ßar a ler os dados.
        # Pode ser um inteiro N (pula as primeiras N linhas) ou uma lista de √≠ndices de linha a pular.
        # Valor: 1 -> Pula a primeira linha (√≠ndice 0), que √© a linha de coment√°rio.
        # Nota: Ap√≥s pular linhas, o Pandas procura o cabe√ßalho na pr√≥xima linha,
        #       a menos que o par√¢metro 'header' especifique outra linha ou seja None.
        #       Neste caso, ap√≥s pular 1 linha, a linha seguinte ("ID;Data;Valor;Cidade")
        #       ser√° corretamente identificada como cabe√ßalho (comportamento padr√£o header=0 relativo ao in√≠cio da leitura).
        skiprows=1,

        # --- Par√¢metro 4: comment ---
        # Caractere que, se aparecer no in√≠cio de uma linha (ap√≥s espa√ßos em branco),
        # indica que o restante dessa linha √© um coment√°rio e deve ser ignorado.
        # Valor: '#' -> Linhas come√ßando com '#' ser√£o ignoradas.
        #        Embora a primeira linha j√° tenha sido pulada por `skiprows`,
        #        este par√¢metro protegeria contra coment√°rios em outras partes do arquivo.
        comment='#',

        # --- Par√¢metro 5: skipfooter ---
        # N√∫mero de linhas a ignorar no *final* do arquivo.
        # Valor: 1 -> Ignora a √∫ltima linha ("Rodap√© a ser ignorado").
        # Importante: Este par√¢metro s√≥ funciona se o 'engine' for 'python'.
        skipfooter=1,

        # --- Par√¢metro 6: engine ---
        # Motor de an√°lise (parser engine) a ser usado. Op√ß√µes: 'c', 'python', 'pyarrow'.
        # Valor: 'python' -> Necess√°rio para usar 'skipfooter'. O motor 'python' √© mais
        #          flex√≠vel, mas geralmente mais lento que o motor 'c' (padr√£o).
        #          O motor 'c' √© otimizado para velocidade, mas tem menos recursos.
        #          'pyarrow' √© uma op√ß√£o mais recente e r√°pida para certos cen√°rios.
        engine='python',

        # --- Par√¢metro 7: thousands ---
        # Caractere a ser interpretado como separador de milhares ao converter strings em n√∫meros.
        # Valor: '.' -> Reconhece o ponto como separador de milhar (ex: "1.250").
        thousands='.',

        # --- Par√¢metro 8: decimal ---
        # Caractere a ser interpretado como separador decimal ao converter strings em n√∫meros.
        # Valor: ',' -> Reconhece a v√≠rgula como separador decimal (ex: "250,50").
        # Nota: 'thousands' e 'decimal' s√£o essenciais para ler corretamente n√∫meros
        #       em formatos n√£o padr√£o (ex: formato monet√°rio brasileiro ou europeu).
        decimal=',',

        # --- Par√¢metro 9: na_values ---
        # String ou lista de strings adicionais que devem ser reconhecidas como NaN (Not a Number / Ausente).
        # Valor: ['?'] -> A string exata "?" ser√° interpretada como um valor ausente (NaN).
        #        Por padr√£o, o Pandas j√° reconhece strings como '', '#N/A', 'NaN', 'NULL', etc.
        #        Use `keep_default_na=False` se quiser *apenas* os valores em `na_values`.
        na_values=['?'],

        # --- Par√¢metro 10: parse_dates ---
        # Tenta converter colunas especificadas para o tipo datetime do Pandas.
        # Valor: ['Data'] -> Instrui o Pandas a tentar analisar a coluna chamada 'Data' como data/hora.
        #        Pode ser True (tenta analisar o √≠ndice), uma lista de nomes/√≠ndices de colunas,
        #        ou um dicion√°rio para combinar m√∫ltiplas colunas em uma data.
        #        Para performance em arquivos grandes, considere converter ap√≥s a leitura com
        #        `pd.to_datetime(df['Data'], format='%Y-%m-%d')` se o formato for conhecido.
        parse_dates=['Data']

        # (Outros par√¢metros comuns e √∫teis n√£o usados aqui incluem:
        #  header=None -> Se n√£o houver linha de cabe√ßalho.
        #  names=['A','B','C'] -> Fornece nomes de colunas (√∫til com header=None).
        #  index_col='ID' -> Usa a coluna 'ID' como √≠ndice do DataFrame.
        #  usecols=['ID','Valor'] -> L√™ apenas um subconjunto de colunas (economiza mem√≥ria).
        #  dtype={'ID':int, 'Valor':float} -> Especifica o tipo de dados para colunas (economia/precis√£o).
        #  encoding='latin-1' -> Especifica a codifica√ß√£o do arquivo (importante para acentos, etc.).
        #  nrows=100 -> L√™ apenas as N primeiras linhas de dados.
        # )
    )

    # Imprime o DataFrame resultante da leitura
    print("\nLeitura CSV Complexo (DataFrame resultante):")
    print(df_csv_complexo)

    # Imprime informa√ß√µes sobre o DataFrame, incluindo tipos de dados por coluna,
    # contagem de valores n√£o nulos e uso de mem√≥ria.
    # √ötil para verificar se a leitura (tipos, NaNs, datas) funcionou como esperado.
    print("\nInforma√ß√µes do DataFrame (.info()):")
    df_csv_complexo.info()

except Exception as e:
    # Captura e informa qualquer erro que possa ocorrer durante a leitura
    print(f"\nErro ao ler o CSV simulado: {e}")
    # Um problema comum poderia ser um FileNotFoundError se estivesse lendo um arquivo real,
    # ou um erro de parsing se os par√¢metros n√£o correspondessem ao formato dos dados.
    # Neste caso, como usamos StringIO, erros seriam mais prov√°veis por par√¢metros incorretos.

print("-" * 50) # Separador visual

# ==============================================================================
# Leitura em Chunks (Conceito para Arquivos Grandes)
# ==============================================================================

# --- Par√¢metro: chunksize ---
# Se voc√™ fornecer um valor inteiro para `chunksize=N`, a fun√ß√£o `pd.read_csv`
# n√£o retorna um DataFrame completo, mas sim um *iterador*.
# Cada vez que voc√™ itera sobre ele, ele l√™ e retorna as pr√≥ximas N linhas
# do arquivo como um DataFrame (o "chunk").
# Objetivo: Processar arquivos que s√£o muito grandes para serem carregados
#           inteiramente na mem√≥ria RAM. Permite processar o arquivo em peda√ßos.

print("\nLeitura CSV em Chunks (exemplo conceitual - c√≥digo comentado):")

# O c√≥digo abaixo est√° comentado porque requer um arquivo real chamado 'arquivo_grande.csv'.
# Ele demonstra a sintaxe e o loop para processar chunks.
"""
try:
    # Define o tamanho do chunk (quantas linhas ler por vez)
    tamanho_chunk = 10000

    # Cria o iterador. Nenhum dado √© lido neste ponto ainda.
    chunk_iter = pd.read_csv('arquivo_grande.csv', chunksize=tamanho_chunk)
                             # Pode incluir outros par√¢metros como sep, header, etc.

    print(f"Iniciando leitura de 'arquivo_grande.csv' em chunks de {tamanho_chunk} linhas...")

    # Lista para armazenar resultados processados de cada chunk (exemplo)
    resultados_processados = []

    # Loop sobre o iterador. A cada itera√ß√£o, l√™ o pr√≥ximo chunk.
    for i, chunk_df in enumerate(chunk_iter):
        # chunk_df √© um DataFrame contendo as linhas do chunk atual
        print(f"Processando chunk {i+1} com shape {chunk_df.shape}...")

        # --- Realize aqui o processamento desejado para cada chunk ---
        # Exemplo: Filtrar dados, calcular estat√≠sticas, transformar colunas, etc.
        # media_valor_chunk = chunk_df['Valor'].mean()
        # resultados_processados.append(media_valor_chunk)
        # -------------------------------------------------------------

        # Libera a mem√≥ria do chunk atual explicitamente se necess√°rio (opcional)
        # del chunk_df

    print("Processamento de todos os chunks conclu√≠do.")
    # Se agregou resultados, pode combin√°-los agora:
    # media_total = pd.Series(resultados_processados).mean()
    # print(f"Resultado agregado (exemplo: m√©dia total): {media_total}")

except FileNotFoundError:
    print("Erro: O arquivo 'arquivo_grande.csv' n√£o foi encontrado para demonstrar chunking.")
except Exception as e:
    print(f"Erro durante o processamento em chunks: {e}")
"""
# Mensagem indicando que o c√≥digo acima √© um exemplo conceitual.
print("(O c√≥digo de exemplo para chunking est√° comentado pois requer um arquivo real)")
print("-" * 50)

print("\n--- Fim da Se√ß√£o 4.1: Leitura de Dados com pd.read_csv ---")

# --- Resumo e Pontos Chave ---
# - `pd.read_csv` √© extremamente flex√≠vel devido aos seus numerosos par√¢metros.
# - Entender par√¢metros como `sep`, `header`, `skiprows`, `skipfooter`, `comment`, `decimal`, `thousands`, `na_values`, `parse_dates`, e `encoding` √© crucial para ler dados "reais" corretamente.
# - O par√¢metro `engine` ('c' vs 'python') afeta performance e disponibilidade de recursos (como `skipfooter`).
# - `StringIO` √© √∫til para testar `read_csv` com dados em strings.
# - `chunksize` √© a solu√ß√£o para ler arquivos que excedem a mem√≥ria dispon√≠vel.
# - Sempre verifique o DataFrame resultante com `.head()`, `.info()`, `.describe()` para garantir que a leitura ocorreu como esperado.
````

````
# -*- coding: utf-8 -*-
"""
Este bloco de c√≥digo demonstra como interagir com um banco de dados SQLite
usando a biblioteca `sqlite3` (padr√£o do Python) e como ler os resultados
de uma consulta SQL diretamente para um DataFrame Pandas usando `pd.read_sql_query`.
Inclui explica√ß√µes detalhadas sobre as fun√ß√µes, sintaxe e par√¢metros.
"""

# Importa a biblioteca sqlite3, que fornece a interface para bancos de dados SQLite.
# SQLite √© um banco de dados leve, baseado em arquivo, muito comum e que j√° vem com Python.
import sqlite3

# Importa a biblioteca Pandas (assume-se que j√° foi importada anteriormente).
# 'pd' √© o alias convencional.
import pandas as pd

print("\n--- 4.x Leitura de Dados de Banco SQL (SQLite) ---")
# (A numera√ß√£o 4.x √© gen√©rica, ajuste conforme a estrutura do seu documento)

# ==============================================================================
# Conex√£o e Prepara√ß√£o do Banco de Dados SQLite (Exemplo em Mem√≥ria)
# ==============================================================================

# Tentaremos executar as opera√ß√µes de banco de dados dentro de um bloco try...finally
# para garantir que a conex√£o seja fechada mesmo se ocorrer um erro.
conn = None # Inicializa a vari√°vel de conex√£o fora do try
try:
    # --- Estabelecendo a Conex√£o ---
    # Fun√ß√£o: sqlite3.connect(database_path)
    # Objetivo: Criar ou conectar a um banco de dados SQLite.
    # Par√¢metros:
    #   database_path (str): O caminho para o arquivo do banco. Se n√£o existir, ser√° criado.
    #       Valor ':memory:': Cria um banco de dados vol√°til *apenas na mem√≥ria RAM*.
    #           Ele √© muito r√°pido e √∫til para exemplos, testes ou dados tempor√°rios,
    #           pois desaparece completamente quando a conex√£o √© fechada.
    # Retorna: Um objeto Connection que representa a conex√£o com o banco.
    conn = sqlite3.connect(':memory:')
    print("Conex√£o com banco de dados SQLite em mem√≥ria estabelecida.")

    # --- Executando Comandos SQL (DDL e DML) ---
    # M√©todo: connection.execute(sql_statement, parameters=None)
    # Objetivo: Executa uma √∫nica instru√ß√£o SQL no banco de dados.
    #          √â um m√©todo conveniente que cria um cursor, executa e retorna o cursor.
    # Par√¢metros:
    #   sql_statement (str): A instru√ß√£o SQL a ser executada.
    #   parameters (tuple/dict, opcional): Usado para passar valores de forma segura
    #             em instru√ß√µes que usam placeholders (como SELECT, INSERT, UPDATE).
    #             N√£o tipicamente usado para DDL (CREATE, ALTER, DROP).

    # 1. Criar Tabela (DDL - Data Definition Language)
    #    Define a estrutura da tabela 'funcionarios'.
    #    Tipos de dados SQLite comuns: INTEGER, REAL (ponto flutuante), TEXT, BLOB, NULL.
    create_table_sql = "CREATE TABLE funcionarios (id INTEGER PRIMARY KEY, nome TEXT, salario REAL)"
    print(f"Executando DDL: {create_table_sql}")
    conn.execute(create_table_sql)

    # 2. Inserir Dados (DML - Data Manipulation Language)
    #    Adiciona duas linhas √† tabela 'funcionarios'.
    insert_sql_1 = "INSERT INTO funcionarios (id, nome, salario) VALUES (1, 'Alice', 5000.0)"
    insert_sql_2 = "INSERT INTO funcionarios (id, nome, salario) VALUES (?, ?, ?)" # Usando placeholders
    params_insert_2 = (2, 'Bob', 6500.5)
    print(f"Executando DML (INSERT 1): {insert_sql_1}")
    conn.execute(insert_sql_1)
    print(f"Executando DML (INSERT 2): {insert_sql_2} com params {params_insert_2}")
    conn.execute(insert_sql_2, params_insert_2) # Passando par√¢metros de forma segura

    # --- Commit das Altera√ß√µes ---
    # M√©todo: connection.commit()
    # Objetivo: Salva permanentemente todas as altera√ß√µes feitas desde o √∫ltimo commit
    #          (ou desde o in√≠cio da conex√£o se for o primeiro). Essencial para que
    #          CREATE TABLE e INSERT/UPDATE/DELETE tenham efeito duradouro no banco.
    conn.commit()
    print("Altera√ß√µes commitadas no banco de dados.")

    # ==============================================================================
    # Leitura de Dados do Banco com Pandas
    # ==============================================================================
    print("\nLeitura de Banco SQL (SQLite) usando pd.read_sql_query:")

    # --- Definindo a Query SQL e Par√¢metros ---
    # Query para selecionar funcion√°rios com sal√°rio acima de um certo valor.
    # O '?' √© um placeholder (estilo 'qmark' padr√£o do sqlite3) que ser√°
    # substitu√≠do de forma segura pelo valor fornecido no par√¢metro `params`.
    query = "SELECT * FROM funcionarios WHERE salario > ?"
    print(f"Query SQL a ser executada: {query}")

    # Par√¢metros para a query. Deve ser uma sequ√™ncia (tupla ou lista) com
    # tantos elementos quantos forem os placeholders '?'.
    # A v√≠rgula ap√≥s 6000 √© crucial para criar uma tupla de um elemento.
    # *** Seguran√ßa: Usar `params` √© a forma correta e segura de passar valores
    # para queries SQL, prevenindo ataques de SQL Injection. NUNCA insira
    # valores diretamente na string da query usando formata√ß√£o de string (f-string, %)
    # se esses valores vierem de fontes externas ou n√£o confi√°veis. ***
    params_sql = (6000,)
    print(f"Par√¢metros para a query: {params_sql}")

    # --- Usando pd.read_sql_query ---
    # Fun√ß√£o: pd.read_sql_query(sql, con, index_col=None, coerce_float=True, params=None, parse_dates=None, chunksize=None)
    # Objetivo: Executar uma consulta SQL e ler os resultados diretamente para um DataFrame Pandas.
    # Par√¢metros Chave:
    #   sql (str): A string contendo a consulta SQL (pode ter placeholders).
    #   con (object): O objeto de conex√£o com o banco de dados (DBAPI2 connection - como `conn` do sqlite3)
    #                 ou uma engine/string de conex√£o SQLAlchemy.
    #   params (list, tuple ou dict, opcional): Sequ√™ncia ou dicion√°rio de par√¢metros
    #              para passar √† consulta SQL de forma segura, substituindo os placeholders.
    #   index_col (str ou list, opcional): Coluna(s) do resultado SQL a ser(em) usada(s) como √≠ndice do DataFrame.
    #   parse_dates (list ou dict, opcional): Coluna(s) a serem interpretadas como datas.
    # Retorna: Um DataFrame Pandas com os resultados da consulta.

    df_sqlite = pd.read_sql_query(query, conn, params=params_sql)

    # Imprime o DataFrame resultante
    print("\nDataFrame resultante da query:")
    print(df_sqlite)
    print("\nTipos de dados do DataFrame:")
    df_sqlite.info() # Mostra que Pandas inferiu os tipos corretamente do SQL

except sqlite3.Error as e:
    # Captura erros espec√≠ficos do sqlite3 (ex: erro de sintaxe SQL)
    print(f"\nErro SQLite ocorrido: {e}")
except Exception as e:
    # Captura outros erros gen√©ricos
    print(f"\nUm erro inesperado ocorreu: {e}")

finally:
    # --- Fechando a Conex√£o ---
    # O bloco `finally` garante que este c√≥digo seja executado SEMPRE,
    # quer o bloco `try` tenha sido bem-sucedido ou tenha gerado um erro.
    if conn: # Verifica se a conex√£o foi estabelecida com sucesso antes de tentar fechar
        # M√©todo: connection.close()
        # Objetivo: Fecha a conex√£o com o banco de dados, liberando recursos.
        #          Para bancos :memory:, isso destr√≥i o banco.
        #          Para bancos em arquivo, garante que tudo foi escrito e desbloqueia o arquivo.
        conn.close()
        print("\nConex√£o com o banco de dados fechada no bloco finally.")

print("-" * 50)

# ==============================================================================
# Observa√ß√£o sobre Gerenciador de Contexto (`with`) - M√©todo Preferido
# ==============================================================================
# A forma mais idiom√°tica e segura de trabalhar com conex√µes `sqlite3` (e muitos
# outros recursos que precisam ser fechados) em Python √© usando um gerenciador
# de contexto (`with` statement). Ele lida automaticamente com o commit/rollback
# e o fechamento da conex√£o.

print("\nExemplo usando 'with' (m√©todo recomendado):")
try:
    # O 'with' garante que conn_with ser√° fechado no final, e lida com commits/rollbacks.
    with sqlite3.connect(':memory:') as conn_with:
        conn_with.execute("CREATE TABLE exemplo_with (id INT, nome TEXT)")
        conn_with.execute("INSERT INTO exemplo_with VALUES (?, ?)", (10, 'Com With'))
        # Commit √© geralmente autom√°tico ao sair do 'with' sem erro.
        # Rollback √© autom√°tico se ocorrer um erro dentro do 'with'.

        df_with = pd.read_sql_query("SELECT * FROM exemplo_with", conn_with)
        print("DataFrame lido dentro do 'with':")
        print(df_with)
    # A conex√£o conn_with j√° est√° fechada aqui.
    print("Bloco 'with' conclu√≠do, conex√£o fechada automaticamente.")
except Exception as e:
    print(f"Erro no exemplo com 'with': {e}")
    # A conex√£o ainda seria fechada automaticamente.

print("-" * 50)
print("--- Fim da Se√ß√£o de Leitura SQL com SQLite ---")
````

* **.to_sql()**: Salva DataFrame em tabela SQL.¬≥‚Å∂ Requer SQLAlchemy.‚Åµ¬≥, ‚Åµ‚Å¥
    * Par√¢metros detalhados: `name` (tabela), `con` (conex√£o), `schema`, `if_exists` ('fail', 'replace', 'append'), `index`, `index_label`, `chunksize` (para inser√ß√£o em lotes), `dtype` (dicion√°rio de tipos SQLAlchemy), `method` ('multi' para inser√ß√£o mais r√°pida com alguns dialetos).‚Åµ¬≥, ‚Åµ‚Å¥

````
# Exemplo conceitual: Salvar DataFrame no banco SQL com tipo de dado definido

# Importa o criador de engine (conex√£o) e tipos de dados espec√≠ficos do SQLAlchemy
from sqlalchemy import create_engine, types

# Cria uma engine de conex√£o com banco SQLite local (um arquivinho .db simples)
engine_db = create_engine('sqlite:///meu_banco_escrita.db')

# Copia um DataFrame existente (para evitar alterar o original sem querer)
df_para_sql = df_pessoas.copy()

# Dicion√°rio mapeando colunas do DataFrame para tipos SQL
dtype_map = {
    'Nome': types.VARCHAR(100),  # Coluna 'Nome' ser√° texto com limite de 100 caracteres
    'Idade': types.INTEGER(),    # Coluna 'Idade' ser√° um inteiro
    'Cidade': types.TEXT()       # Coluna 'Cidade' ser√° texto livre (sem limite)
}

try:
    # Exporta o DataFrame para uma tabela SQL com controle de tipos
    df_para_sql.to_sql(
        'tabela_pessoas',       # Nome da tabela no banco de dados
        con=engine_db,          # Conex√£o do banco (a engine que criamos)
        if_exists='replace',    # Se a tabela j√° existir, apaga e recria (cuidado!)
        index=False,            # N√£o exporta o √≠ndice do DataFrame como coluna na tabela
        dtype=dtype_map,        # Aplica os tipos definidos para cada coluna
        method='multi'          # Envia v√°rias linhas por vez (batch), melhora performance em alguns casos
    )
    print("\nDataFrame salvo em tabela SQL 'tabela_pessoas'.")

except Exception as e:
    # Captura qualquer erro que ocorrer durante o processo e imprime
    print(f"Erro ao salvar em SQL: {e}")

finally:
    # Encerra a conex√£o com o banco, liberando recursos
    if 'engine_db' in locals():
        engine_db.dispose()
````

### üß† Extras que vale saber:

- `method='multi'`: Nem todo banco suporta isso, mas quando suporta, √© bala ‚Äî ele insere **v√°rias linhas em uma s√≥ query**. Bem mais r√°pido.
- `types.VARCHAR`, `types.INTEGER`: S√£o classes da SQLAlchemy que definem como o SQL vai criar as colunas.
- `create_engine('sqlite:///...')`: Para SQLite local. Mas pode ser PostgreSQL, MySQL, etc. (ex: `'postgresql://user:senha@host:porta/db'`)

* **Sele√ß√£o de Colunas:** `df['col']` (Series), `df[['col1', 'col2']]` (DataFrame), `df.col` (menos recomendado).¬≤‚Å∏, ¬≤‚Åπ
* **Sele√ß√£o de Linhas:** Slicing `df[start:stop]`. **Preferir `.loc`/`.iloc`**.¬≤‚Å∏
* **Sele√ß√£o Baseada em R√≥tulos (`.loc`):** Usa r√≥tulos de √≠ndice/coluna. Slice final inclusivo. Aceita booleanos e callables.¬≤‚Å∏, ¬≥‚Å∞

````
print("\n--- 6. Sele√ß√£o e Indexa√ß√£o Detalhada ---")
import pandas as pd

# Criar um DataFrame com colunas necess√°rias
df = pd.DataFrame({
    'ID': [0, 1, 2, 3, 4, 5, 6],
    'Genero': ['F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'Idade': [25, 32, 21, 45, 28, 38, 22],
    'Pontuacao_Gasto': [88.5, 72.0, 91.2, 60.4, 78.0, 55.5, 84.3]
})

print("DataFrame original:")
print(df)

# Pega as 5 primeiras linhas e define a coluna 'ID' como √≠ndice (importante pro .loc funcionar com r√≥tulos)
df_loc_ex = df.head(5).set_index('ID')
print("\nDataFrame para exemplos .loc/.iloc:\n", df_loc_ex)

# Acessa a linha cujo √≠ndice (ID) √© 2 ‚Üí retorna uma Series
print("\n.loc[2]:\n", df_loc_ex.loc[2])

# Acessa linhas 3 e 1 (por ID) e colunas 'Pontuacao_Gasto' e 'Genero' (ordem arbitr√°ria)
print("\n.loc[[3, 1], ['Pontuacao_Gasto', 'Genero']]:\n", df_loc_ex.loc[[3, 1], ['Pontuacao_Gasto', 'Genero']])

# Slice de linhas de ID 1 at√© 3 (inclusive) e colunas de 'Idade' at√© 'Pontuacao_Gasto' (tamb√©m inclusivo)
print("\n.loc[1:3, 'Idade':'Pontuacao_Gasto']:\n", df_loc_ex.loc[1:3, 'Idade':'Pontuacao_Gasto'])

# Filtra linhas onde a idade √© menor que 30
print("\n.loc[df_loc_ex['Idade'] < 30]:\n", df_loc_ex.loc[df_loc_ex['Idade'] < 30])
````

* **Sele√ß√£o Baseada em Posi√ß√£o (`.iloc`):** Usa posi√ß√µes inteiras (base 0). Slice final exclusivo. Aceita booleanos (`.values`) e callables.¬π, ¬≤‚Å∏

````
# ------------------------
# EXEMPLOS COM .iloc
# ------------------------

# Mostra a PRIMEIRA LINHA do DataFrame (posi√ß√£o 0)
print("\n.iloc[0]:\n", df_loc_ex.iloc[0])

# Seleciona:
# - Linhas nas posi√ß√µes 1 e 3
# - Colunas nas posi√ß√µes 0 e 2 (ou seja, 2¬™ e 4¬™ colunas do DataFrame)
print("\n.iloc[[1, 3], [0, 2]]:\n", df_loc_ex.iloc[[1, 3], [0, 2]])

# Slice de:
# - Linhas da posi√ß√£o 1 at√© a 3 (exclusivo da 4)
# - Colunas da posi√ß√£o 0 at√© a 2 (exclusivo da 3)
# Equivalente a pegar 2¬™, 3¬™ e 4¬™ linha e 1¬™, 2¬™ e 3¬™ coluna
print("\n.iloc[1:4, :3]:\n", df_loc_ex.iloc[1:4, :3])

# Filtro booleano com .iloc exige um array do numpy (sem index, s√≥ True/False na ordem)
# Aqui pegamos todas as linhas onde a Pontua√ß√£o √© maior que 50
condicao_iloc = (df_loc_ex['Pontuacao_Gasto'] > 50).values  # Isso gera um array como [True, True, True, ...]
print("\n.iloc[condicao_booleana.values]:\n", df_loc_ex.iloc[condicao_iloc])
````

* **Indexa√ß√£o Booleana:** Filtrar com condi√ß√µes (`&`, `|`, `~`, `.isin()`, `.str.contains()`, etc.).‚Å∂‚Å∂

````
print("\nExemplos de Indexa√ß√£o Booleana:")

import pandas as pd

# Criando um DataFrame de exemplo
df = pd.DataFrame({
    'ID': range(1, 11),  # IDs de 1 a 10
    'Genero': ['F', 'M', 'F', 'M', 'F', 'F', 'M', 'F', 'M', 'F'],  # G√™neros variados
    'Pontuacao_Gasto': [75, 65, 85, 40, 90, 30, 60, 95, 70, 80],  # Pontua√ß√£o de gasto
    'Renda_Anual_Mil': [95, 88, 120, 45, 100, 55, 70, 150, 85, 110],  # Renda em milhares
    'Categoria_Pref': [
        'Eletr√¥nicos', 'Roupas', 'Livros', 'Alimentos',
        'Viagens', 'Eletr√¥nicos', 'Livros', 'Viagens',
        'Roupas', 'Eletr√¥nicos'  # Prefer√™ncia de categoria
    ],
    'Idade': [25, 32, 29, 45, 37, 22, 34, 39, 28, 41]  # Idades diversas
})

# -------------------------------------------
# 1. Filtro complexo usando operadores l√≥gicos
# -------------------------------------------

# Cria um DataFrame filtrado com base nas seguintes condi√ß√µes:
# - G√™nero feminino ('F')
# - E (Pontua√ß√£o de Gasto > 70 OU Renda Anual acima de 90 mil)
filtro_complexo = df[
    (df['Genero'] == 'F') &  # Verifica se a pessoa √© do g√™nero feminino
    (
        (df['Pontuacao_Gasto'] > 70) |  # OU: se a pontua√ß√£o de gasto for maior que 70
        (df['Renda_Anual_Mil'] > 90)    # OU: se a renda anual for maior que 90 mil
    )
]

# Mostra as colunas relevantes para esse filtro, usando .head() para exibir s√≥ os primeiros resultados
print("Feminino E (Score>70 OU Renda>90k):\n",
      filtro_complexo[['ID', 'Genero', 'Pontuacao_Gasto', 'Renda_Anual_Mil']].head())


# ---------------------------------------------------------
# 2. Filtro com .isin() para categorias espec√≠ficas (OU)
# ---------------------------------------------------------

# Define uma lista com as categorias de interesse
categorias_interesse = ['Eletr√¥nicos', 'Livros']

# Filtra o DataFrame para manter apenas as linhas onde 'Categoria_Pref'
# cont√©m um dos valores da lista 'categorias_interesse'
filtro_isin = df[df['Categoria_Pref'].isin(categorias_interesse)]

# Exibe os primeiros resultados contendo apenas as colunas 'ID' e 'Categoria_Pref'
print(f"\nClientes com Categoria em {categorias_interesse}:\n",
      filtro_isin[['ID', 'Categoria_Pref']].head())


# ------------------------------------------
# 3. Filtro com .between() (intervalo inclusivo)
# ------------------------------------------

# Filtra os clientes cuja idade est√° entre 30 e 40 anos, inclusive
filtro_between = df[df['Idade'].between(30, 40)]  # between inclui os limites inferior e superior

# Exibe os primeiros clientes que atendem √† condi√ß√£o, mostrando ID e idade
print("\nClientes com idade entre 30 e 40:\n",
      filtro_between[['ID', 'Idade']].head())
````

* **`.at` / `.iat`:** Acesso/modifica√ß√£o escalar r√°pido por r√≥tulo/posi√ß√£o.

**7. Manipula√ß√£o de Dados** (Expandido)

(Exemplos de cria√ß√£o/dele√ß√£o de colunas, `.apply`, `.map`, `.rename`, `.drop` j√° na vers√£o anterior, agora adicionamos `.assign`, `.applymap`)

````
import pandas as pd
import numpy as np

print("\n--- 7. Manipula√ß√£o de Dados (Adicional) ---")

# Simulando um DataFrame limpo
df_clean = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Idade': [25, 40, 30, 35, 28],
    'Genero': ['M', 'F', 'M', 'F', 'M'],
    'Categoria_Pref': ['Viagens', 'Eletr√¥nicos', 'Livros', 'Vestu√°rio', 'Vinhos'],
    'Pontuacao_Gasto': [60, 90, 70, 85, 75],
    'Renda_Anual_Mil': [70, 120, 80, 95, 65]
})

df_manip = df_clean.copy()

# --- .assign() ---
# Cria novas colunas de forma encadeada sem modificar o DataFrame original
# Sintaxe: df.assign(NOME_COLUNA=fun√ß√£o_ou_valor)
# A fun√ß√£o recebe o DataFrame atual como argumento (com lambda)
df_assign = df_manip.assign(
    # np.log1p(x) ‚Üí log(1 + x), evita problema com log(0)
    Renda_Log = lambda x: np.log1p(x['Renda_Anual_Mil']),

    # Normaliza score para o intervalo [0,1]
    Score_Normalizado = lambda x: (
        (x['Pontuacao_Gasto'] - x['Pontuacao_Gasto'].min()) /
        (x['Pontuacao_Gasto'].max() - x['Pontuacao_Gasto'].min())
    ),

    # Multiplica Idade pela Pontua√ß√£o ‚Üí Feature Engineering puro
    Idade_x_Score = lambda x: x['Idade'] * x['Pontuacao_Gasto']
)

print("\nNovas colunas com .assign():")
print(df_assign[['Renda_Log', 'Score_Normalizado', 'Idade_x_Score']].head().round(2))


# --- .apply() ---
# Aplica uma fun√ß√£o a cada coluna (axis=0) ou linha (axis=1)
# Muito usado para aplicar transforma√ß√µes em massa
# Sintaxe: df.apply(funcao, axis=0 ou 1)
# axis=0 ‚Üí coluna por coluna (padr√£o)
# axis=1 ‚Üí linha por linha (cada linha vira uma Series)

def log_transform(coluna):
    # Checa se a coluna √© num√©rica (evita erro ao aplicar log em strings)
    if pd.api.types.is_numeric_dtype(coluna):
        return np.log1p(coluna)
    return coluna

df_log = df_clean.apply(log_transform, axis=0)
print("\nDataFrame ap√≥s aplicar log1p a colunas num√©ricas (.apply axis=0):")
print(df_log.head().round(2))


# --- .applymap() ---
# Aplica uma fun√ß√£o c√©lula por c√©lula (elemento a elemento)
# CUIDADO: s√≥ funciona com DataFrames (n√£o com Series)
# Ideal para aplicar fun√ß√µes simples como limpeza de strings
# OBS: Est√° caindo em desuso ‚Äî prefira .map() para Series e .apply() com axis

def para_maiuscula(valor):
    # Se for string, converte pra mai√∫scula
    if isinstance(valor, str):
        return valor.upper()
    return valor

df_upper = df_clean.applymap(para_maiuscula)
print("\nDataFrame ap√≥s .applymap(para_maiuscula):")
print(df_upper.head())
````


### TL;DR das fun√ß√µes:

| Fun√ß√£o        | O que faz | Quando usar | Exemplo r√°pido |
|---------------|-----------|-------------|----------------|
| `assign()`    | Cria colunas novas sem modificar o DF original | Feature Engineering encadeada | `df.assign(nova_col=x['col1']*2)` |
| `apply()`     | Aplica fun√ß√£o por coluna (ou linha) | Transforma√ß√µes mais complexas | `df.apply(funcao, axis=0)` |
| `applymap()`  | Aplica fun√ß√£o em cada c√©lula individual | Strings ou limpeza geral | `df.applymap(str.strip)` |
| `np.log1p(x)` | Calcula `log(1 + x)` | Quando tem zeros no dado | `np.log1p(df['col'])` |
| `is_numeric_dtype()` | Testa se coluna √© num√©rica | Valida√ß√£o antes de aplicar c√°lculo | `if is_numeric_dtype(col):` |


**8. Grouping e Agrega√ß√£o (`.groupby()`)** (Expandido)

(Exemplos b√°sicos de `.groupby().agg()` j√° na vers√£o anterior, adicionamos `.transform()`, `.filter()`)

````
import pandas as pd
import numpy as np

print("\n--- 8. Grouping e Agrega√ß√£o (Adicional) ---")

# Criando um DataFrame de exemplo para funcionar com o c√≥digo
df_final = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5, 6],
    'Regiao': ['Norte', 'Sul', 'Sul', 'Norte', 'Sudeste', 'Sudeste'],
    'Genero': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Idade': [25, 35, 45, 30, 40, 29],
    'Score': [50, 60, 70, 55, 80, 65],
    'Renda_k_Ano': [60, 80, 100, 70, 90, 85]
})

# Copiando para manter o original intacto
df_agg = df_final.copy()

# ===============================
# .transform(): Agrupa e retorna um resultado com o MESMO n√∫mero de linhas do DataFrame original.
# Aqui, agrupamos por 'Regiao' e aplicamos a m√©dia ('mean') na coluna 'Renda_k_Ano'.
# O resultado ser√° uma nova coluna onde cada linha recebe a m√©dia da sua pr√≥pria regi√£o.
# Isso √© √∫til quando queremos agregar mas manter o alinhamento com os dados originais.

df_agg['Renda_Media_Regiao'] = df_agg.groupby('Regiao')['Renda_k_Ano'].transform('mean')

print("\nDataFrame com Renda M√©dia da Regi√£o (.transform()):")
print(df_agg[['Regiao', 'Renda_k_Ano', 'Renda_Media_Regiao']].head().round(1))


# ===============================
# .filter(): Serve para eliminar grupos inteiros com base em uma condi√ß√£o.
# Neste exemplo, usamos lambda para manter SOMENTE as regi√µes cuja m√©dia da coluna 'Score' > 55.
# A fun√ß√£o √© aplicada a cada grupo (grupo = subset do DataFrame com a mesma 'Regiao').
# Retorna um DataFrame somente com os grupos que passam no filtro.

df_filtrado_grupo = df_agg.groupby('Regiao').filter(lambda grupo: grupo['Score'].mean() > 55)

print("\nDataFrame filtrado por Grupo (.filter() - Regi√µes com Score M√©dio > 55):")
print("Regi√µes Mantidas:", df_filtrado_grupo['Regiao'].unique())
print(f"Shape Original: {df_agg.shape}, Shape Filtrado: {df_filtrado_grupo.shape}")


# ===============================
# .agg(): Usado com groupby para aplicar fun√ß√µes agregadas a cada grupo.
# Podemos passar:
# - Uma fun√ß√£o simples ('mean', 'sum', etc)
# - Uma lista de fun√ß√µes ['mean', 'median']
# - Uma fun√ß√£o customizada, como faremos agora

# Fun√ß√£o customizada: intervalo interquartil (IQR)
def range_iqr(series):
    q1 = series.quantile(0.25)  # 1¬∫ quartil
    q3 = series.quantile(0.75)  # 3¬∫ quartil
    return q3 - q1              # Diferen√ßa = IQR

# Aplicamos a agrega√ß√£o agrupando por 'Genero' e usando:
# - m√©dia ('mean')
# - mediana ('median')
# - a fun√ß√£o customizada range_iqr definida acima

print("\nAgrega√ß√£o customizada (IQR da Idade por Genero):")
print(df_agg.groupby('Genero')['Idade'].agg(['mean', 'median', range_iqr]))
````

**9. Combina√ß√£o de DataFrames** (Expandido)

(Exemplos b√°sicos de `concat`, `merge` j√° na vers√£o anterior, adicionamos detalhes)

````
import pandas as pd

print("\n--- 9. Combina√ß√£o de DataFrames (Adicional) ---")

# ======================================
# Criando dois DataFrames simples com colunas e √≠ndices diferentes
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']}, index=[0, 1])
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'C': ['C2', 'C3']}, index=[0, 2])  # √çndice 2 e coluna 'C' diferente

# ======================================
# pd.concat(): concatena DataFrames ao longo do eixo 0 (linhas) ou 1 (colunas)
# Par√¢metros principais:
# - objs: lista de DataFrames
# - axis: 0 para empilhar verticalmente (default), 1 para juntar lado a lado
# - join: 'outer' (default) une todas colunas, 'inner' s√≥ mant√©m colunas comuns
# - keys: adiciona um n√≠vel de √≠ndice (MultiIndex) identificando a origem

# Aqui, empilhamos df1 e df2 verticalmente com keys = ['fonte1', 'fonte2']
concat_keys = pd.concat([df1, df2], keys=['fonte1', 'fonte2'])

print("\nConcatena√ß√£o Vertical com Keys (MultiIndex):\n", concat_keys)
# MultiIndex: o primeiro n√≠vel √© 'fonte1' ou 'fonte2', o segundo √© o √≠ndice original (0, 1, etc)

# Podemos acessar as "fatias" do MultiIndex com .loc
print("\nAcessando via MultiIndex (concat_keys.loc['fonte1']):\n", concat_keys.loc['fonte1'])

# ======================================
# Concatena√ß√£o horizontal (lado a lado), unindo pelas linhas (√≠ndices)
# join='inner' mant√©m apenas os √≠ndices que est√£o em AMBOS os DataFrames

concat_h_inner = pd.concat([df1, df2], axis=1, join='inner')
print("\nConcatena√ß√£o Horizontal (join='inner') - Apenas √≠ndices comuns:\n", concat_h_inner)
# Apenas o √≠ndice 0 √© comum entre df1 e df2

# ======================================
# Merge entre dois DataFrames com m√∫ltiplas chaves ('key1' e 'key2')

df_left = pd.DataFrame({
    'key1': ['K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0'],
    'A': ['A0', 'A1', 'A2']
})

df_right = pd.DataFrame({
    'key1': ['K0', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0'],
    'B': ['B0', 'B1', 'B2']
})

print("\nDataFrames para Merge M√∫ltiplo:\ndf_left:\n", df_left)
print("df_right:\n", df_right)

# pd.merge(): semelhante ao SQL JOIN
# Par√¢metros:
# - on: colunas para o merge (pode ser uma lista para m√∫ltiplas)
# - how: tipo de join ('inner', 'left', 'right', 'outer')
# - indicator: True adiciona coluna '_merge' indicando a origem da linha
# - suffixes: sufixos para colunas com o mesmo nome nos dois DataFrames

merge_multi = pd.merge(
    df_left, df_right,
    on=['key1', 'key2'],  # merge por m√∫ltiplas colunas
    how='outer',          # mant√©m todas as combina√ß√µes poss√≠veis
    indicator=True,       # adiciona coluna '_merge'
    suffixes=('_esq', '_dir')  # sufixos se houver nomes duplicados de coluna
)
print("\nMerge Outer com M√∫ltiplas Chaves e Indicator:\n", merge_multi)

# ======================================
# Merge entre √≠ndice de um DataFrame e coluna de outro

# Primeiro, definimos o √≠ndice de df_left como 'key1'
df_left_idx = df_left.set_index('key1')

# Copiamos df_right para manter original
df_right_col = df_right.copy()

# Aqui fazemos o merge entre:
# - o √≠ndice de df_left_idx
# - a coluna 'key1' de df_right_col
# Isso √© √∫til quando temos DataFrames estruturados de forma diferente

merge_idx_col = pd.merge(
    df_left_idx, df_right_col,
    left_index=True,    # usa o √≠ndice de df_left_idx ('key1')
    right_on='key1',    # usa a coluna 'key1' de df_right_col
    how='left',         # todas as linhas da esquerda s√£o mantidas
    suffixes=('_cli', '_comp')
)
print("\nMerge √çndice com Coluna:\n", merge_idx_col)
````

**10. Tratamento de Dados Ausentes (Revis√£o/Expans√£o)**

(Exemplos b√°sicos j√° vistos, adicionamos `limit`, `interpolate`)

````
import pandas as pd
import numpy as np

print("\n--- 10. Tratamento de Dados Ausentes (Adicional) ---")

# ===========================================
# Criando um DataFrame com valores ausentes (NaN)
# np.nan representa um valor "n√£o num√©rico" (Not a Number)
df_miss = pd.DataFrame({
    'A': [1, 2, np.nan, np.nan, 5, np.nan],
    'B': [10, np.nan, 30, 40, np.nan, 60]
})
print("\nDataFrame para Dados Ausentes:\n", df_miss)

# ===========================================
# .fillna(): Preenche valores ausentes (NaN) com uma estrat√©gia
# Par√¢metros √∫teis:
# - method='ffill': "forward fill", copia o valor anterior n√£o-nulo para frente
# - limit=N: s√≥ preenche at√© N NaNs consecutivos
# - value=algum_valor: define um valor fixo para preenchimento
# - inplace=True: altera o pr√≥prio DataFrame (caso contr√°rio retorna um novo)

# Aqui fazemos forward fill com limit=1 ‚Üí s√≥ preenche 1 NaN por vez
print("\nfillna(method='ffill', limit=1):\n", df_miss.fillna(method='ffill', limit=1))

# Resultado:
#   A: preenche o primeiro NaN com o valor acima (2), mas deixa o segundo
#   B: preenche o NaN da linha 1 com 10, mas pula os demais

# ===========================================
# .interpolate(): interpola√ß√£o linear entre valores conhecidos
# Ideal para s√©ries temporais ou dados num√©ricos cont√≠nuos
# Par√¢metros:
# - method='linear' (default): interpola√ß√£o simples entre dois pontos
# - limit: n√∫mero m√°ximo de valores consecutivos a preencher
# - axis: 0 (linha, default) ou 1 (coluna)
# - inplace=True: altera o DataFrame original
# ‚ö†Ô∏è Cuidado: se o primeiro ou √∫ltimo valor for NaN, ele pode n√£o ser preenchido

# Aplicamos interpola√ß√£o na coluna 'A'
df_miss['A_interpolado'] = df_miss['A'].interpolate(method='linear')

print("\nDataFrame com Coluna A Interpolada:\n", df_miss)

# Explica√ß√£o da interpola√ß√£o:
# A = [1.0, 2.0, NaN, NaN, 5.0, NaN]
# A interpolado fica:
# linha 2: 2.0 + (5.0 - 2.0) * 1/3 = 3.0
# linha 3: 2.0 + (5.0 - 2.0) * 2/3 = 4.0
# linha 5: continua NaN pois n√£o h√° um valor depois de 5 para interpolar
````

üß† Dicas ninja:

- `fillna()` √© √≥timo para preencher com m√©dias, mediana, moda ou valores fixos. Exemplo:
  ```python
  df['coluna'].fillna(df['coluna'].mean())
  ```
- `interpolate()` √© melhor para **tend√™ncia num√©rica cont√≠nua**, como s√©ries temporais.
- Evite `fillna(method='ffill')` sem limite em dados com muitos NaNs no in√≠cio ‚Äî pode propagar valores irreais.
- Para apagar linhas ou colunas com muitos NaNs:
  ```python
  df.dropna(axis=0, how='any')   # remove linhas com pelo menos 1 NaN
  df.dropna(axis=1, how='all')   # remove colunas completamente vazias
  ```


````
import pandas as pd
import numpy as np

print("\n--- 10. Tratamento de Dados Ausentes (Avan√ßado) ---")

# ===========================================
# Criando um DataFrame com valores ausentes (NaN)
df_miss = pd.DataFrame({
    'A': [1, 2, np.nan, np.nan, 5, np.nan],
    'B': [10, np.nan, 30, 40, np.nan, 60]
})
print("\nDataFrame para Dados Ausentes:\n", df_miss)

# ===========================================
# .dropna(): Remove linhas ou colunas com valores ausentes (NaN)
# Par√¢metros:
# - axis=0 (linha, default) ou axis=1 (coluna)
# - how='any': remove linhas/colunas com pelo menos um NaN
# - how='all': remove linhas/colunas com TODOS os valores como NaN
# - thresh=N: mant√©m as linhas/colunas com pelo menos N valores n√£o-NaN

# Exemplo: Remover linhas com qualquer NaN
df_drop_any = df_miss.dropna(axis=0, how='any')
print("\nRemovendo Linhas com Qualquer NaN (.dropna(axis=0, how='any')):\n", df_drop_any)

# Exemplo: Remover colunas com TODOS os valores NaN
df_drop_all = df_miss.dropna(axis=1, how='all')
print("\nRemovendo Colunas com Todos os Valores NaN (.dropna(axis=1, how='all')):\n", df_drop_all)

# Exemplo: Manter linhas com pelo menos 2 valores n√£o-NaN
df_drop_thresh = df_miss.dropna(axis=0, thresh=2)
print("\nMantendo Linhas com Pelo Menos 2 Valores N√£o-NaN (.dropna(thresh=2)):\n", df_drop_thresh)

# ===========================================
# .isna() / .notna(): Identifica valores ausentes ou n√£o ausentes
# .isna(): retorna True onde h√° NaN
# .notna(): retorna True onde N√ÉO h√° NaN

# Exemplo de uso de .isna() para identificar NaNs
isna_result = df_miss.isna()
print("\nResultado de .isna() (True para NaN, False para valores presentes):\n", isna_result)

# Exemplo de uso de .notna() para identificar valores presentes
notna_result = df_miss.notna()
print("\nResultado de .notna() (True para valores presentes, False para NaN):\n", notna_result)

# ===========================================
# .mask(): Substitui valores com base em uma condi√ß√£o
# .mask(condi√ß√£o): substitui os valores que atendem √† condi√ß√£o

# Exemplo: Substituir NaNs em 'A' por 0
df_mask = df_miss.copy()
df_mask['A'] = df_mask['A'].mask(df_mask['A'].isna(), 0)
print("\nDataFrame ap√≥s .mask() para substituir NaNs por 0:\n", df_mask)

# ===========================================
# T√©cnicas de Imputa√ß√£o Avan√ßada

# Imputa√ß√£o com valores de outras colunas: aqui usamos o valor m√©dio de 'A' para preencher os NaNs de 'B'
df_imputacao = df_miss.copy()
df_imputacao['B'] = df_imputacao['B'].fillna(df_imputacao['A'].mean())  # Imputa NaNs de 'B' com a m√©dia de 'A'
print("\nDataFrame ap√≥s imputa√ß√£o da Coluna 'B' com a m√©dia de 'A':\n", df_imputacao)

# Imputa√ß√£o usando o m√©todo de interpola√ß√£o para 'B' (aplica a interpola√ß√£o em uma coluna)
df_miss['B_interpolado'] = df_miss['B'].interpolate(method='linear')
print("\nDataFrame ap√≥s interpola√ß√£o linear na Coluna 'B':\n", df_miss)

# Imputa√ß√£o usando KNN (K-Nearest Neighbors) da biblioteca 'sklearn' (exemplo avan√ßado):
from sklearn.impute import KNNImputer

# Criando um imputer com 2 vizinhos mais pr√≥ximos
imputer = KNNImputer(n_neighbors=2)

# Imputando os valores ausentes usando KNN (em todo o DataFrame)
df_knn_imputado = pd.DataFrame(imputer.fit_transform(df_miss), columns=df_miss.columns)
print("\nDataFrame ap√≥s imputa√ß√£o usando KNN:\n", df_knn_imputado)
````

**11. S√©ries Temporais (B√°sico)** (Expandido)

(Exemplo b√°sico j√° visto, adicionamos detalhes e `.shift()`, `.rolling()`)

````
print("\n--- 11. S√©ries Temporais (Adicional) ---")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criando um DataFrame simulado com dados de visitas, renda e pontua√ß√£o
df_clean = pd.DataFrame({
    'ID': range(1, 31),  # 30 visitas fict√≠cias
    'Data_Visita': pd.date_range(start='2024-07-01', periods=30, freq='D'),  # Datas de 1 a 30 de julho de 2024
    'Renda_Anual_Mil': np.random.randint(50, 200, size=30),  # Renda em milhares
    'Pontuacao_Gasto': np.random.randint(0, 100, size=30)  # Pontua√ß√£o de gasto aleat√≥ria
})

# C√≥pia do DataFrame limpo (df_clean), onde assumimos que j√° foi tratado antes
df_ts = df_clean.copy()

# Converte a coluna 'Data_Visita' para o tipo datetime (fundamental para trabalhar com s√©ries temporais!)
# Isso permite realizar filtros, ordena√ß√µes e opera√ß√µes por tempo com efici√™ncia
df_ts['Data_Visita'] = pd.to_datetime(df_ts['Data_Visita'])

# Define a coluna 'Data_Visita' como √≠ndice do DataFrame
# O Pandas trata √≠ndices do tipo datetime de forma especial, permitindo indexa√ß√£o e reamostragem por datas
df_ts.set_index('Data_Visita', inplace=True)

# Ordena o √≠ndice (ou seja, as datas) em ordem crescente
# Isso √© obrigat√≥rio para que opera√ß√µes temporais funcionem corretamente (ex: rolling, shift, etc.)
df_ts.sort_index(inplace=True)

print("\nDataFrame S√©rie Temporal (head):\n", df_ts.head())

# -----------------------------------------
# Sele√ß√£o parcial por data
# Aqui usamos uma feature top do Pandas: sele√ß√£o de datas com strings parciais.
# Por exemplo, '2024-07' pega tudo que for Julho de 2024.
print("\nCompras em 2024-Julho:")
print(df_ts.loc['2024-07'].head())  # ATEN√á√ÉO: Isso s√≥ funciona se o √≠ndice for datetime

# -----------------------------------------
# Reamostragem semanal: agrupar os dados por semana (com in√≠cio na segunda-feira)
# 'W-MON' significa "weekly (semanal), starting on Monday"
# .resample() permite agregar dados por intervalos temporais (di√°rio, mensal, anual, etc)
resample_semanal = df_ts.resample('W-MON').agg(
    Contagem=('ID', 'size'),  # Conta o n√∫mero de visitas (baseado no n√∫mero de IDs por semana)
    Renda_Total_Semana=('Renda_Anual_Mil', 'sum')  # Soma o total de Renda Anual para cada semana
)
print("\nResample Semanal (Contagem e Soma Renda):")
print(resample_semanal.head())

# -----------------------------------------
# Deslocamento temporal com .shift()
# Muito √∫til em s√©ries temporais para criar features como "valor anterior" (lag) ou "valor futuro" (lead)
# Neste exemplo, criamos uma nova coluna com o valor anterior da pontua√ß√£o
df_ts['Score_Anterior'] = df_ts['Pontuacao_Gasto'].shift(1)

# .shift(1): move os valores 1 linha para baixo (lag de 1). A primeira linha vira NaN, pois n√£o tem anterior.
print("\nDataFrame com Score da Visita Anterior (.shift(1)):")
print(df_ts[['Pontuacao_Gasto', 'Score_Anterior']].head())

# -----------------------------------------
# Janela m√≥vel (Rolling Window): c√°lculo de estat√≠sticas em janelas de tempo m√≥veis
# Antes, precisamos garantir que temos **dados cont√≠nuos no tempo**, o que raramente acontece em dados brutos
# Por isso fazemos um .resample() di√°rio, seguido de .ffill() para preencher os valores ausentes

# Reamostragem para di√°rio (D = daily), preenchendo os dias faltantes com o valor anterior (.ffill)
df_ts_diario = df_ts['Renda_Anual_Mil'].resample('D').mean().ffill()

# Calcula m√©dia m√≥vel de 7 dias:
# .rolling(window=7): cria janelas de 7 dias (pode ser baseada em observa√ß√µes ou tempo real)
# .mean(): aplica a m√©dia sobre cada janela
# min_periods=1: exige no m√≠nimo 1 valor para calcular a m√©dia (sen√£o seria NaN at√© preencher 7)
df_ts['Renda_Media_Movel_7D'] = df_ts_diario.rolling(window=7, min_periods=1).mean()

print("\nDataFrame com M√©dia M√≥vel de 7 dias da Renda (.rolling(7).mean()):")
print(df_ts[['Renda_Anual_Mil', 'Renda_Media_Movel_7D']].dropna().head())
# dropna() usado aqui apenas para exibir onde realmente come√ßa a s√©rie preenchida

# -----------------------------------------
# Visualiza√ß√£o com matplotlib
# Import necess√°rio (caso n√£o tenha sido feito antes):
import matplotlib.pyplot as plt

# Plota a s√©rie original interpolada (di√°ria) e a m√©dia m√≥vel de 7 dias
plt.figure(figsize=(12, 5))
df_ts['Renda_Anual_Mil'].plot(label='Renda Original (Di√°ria Interpolada)', alpha=0.5)
df_ts['Renda_Media_Movel_7D'].plot(label='M√©dia M√≥vel 7 Dias', color='red')
plt.title('Renda Anual (Milhares) e M√©dia M√≥vel 7 Dias')
plt.legend()
plt.show()
````

**13. Otimiza√ß√£o de Performance**

Para DataFrames grandes, considere:

* **Tipos de Dados Eficientes:** Use dtypes menores quando poss√≠vel (e.g., `int8`, `int16`, `float32`). Use `pd.Categorical` para strings com pouca variedade (economiza muita mem√≥ria).

````
# === OTIMIZA√á√ÉO DE TIPOS DE DADOS (DTYPES) PARA ECONOMIZAR MEM√ìRIA ===
import pandas as pd
import numpy as np

df_final = pd.DataFrame({
    'Idade': np.random.randint(18, 70, 1000),
    'Pontuacao_Gasto': np.random.randint(0, 100, 1000),
    'Anos_Cliente': np.random.randint(0, 20, 1000),
    'Genero': np.random.choice(['Feminino', 'Masculino'], 1000),
    'Regiao': np.random.choice(['Norte', 'Sul', 'Sudeste'], 1000),
    'Categoria_Pref': np.random.choice(['A', 'B', 'C'], 1000),
    'Faixa_Etaria': np.random.choice(['18-25', '26-35', '36-50', '50+'], 1000)
})

# Criamos uma c√≥pia do DataFrame original para n√£o modificar os dados brutos durante a otimiza√ß√£o.
df_opt = df_final.copy()

# Exibe os tipos de dados originais de cada coluna.
print("\nDtypes Originais:")
print(df_opt.dtypes)

# Calcula o uso total de mem√≥ria do DataFrame em megabytes.
# memory_usage(deep=True) faz uma an√°lise mais precisa, incluindo strings e objetos.
# .sum() soma o uso de todas as colunas.
# Dividimos por 1024^2 para converter bytes em megabytes.
print(f"Uso de mem√≥ria inicial: {df_opt.memory_usage(deep=True).sum() / 1024**2:.2f} MB")


# === CONVERS√ÉO DE INTEIROS PARA TIPOS MENORES ===

# Aqui, reduzimos o tamanho das colunas num√©ricas inteiras.
# Se os valores cabem dentro de um intervalo menor, podemos trocar de tipo.
# Isso economiza bastante mem√≥ria, especialmente em grandes datasets.

# np.int8 = n√∫mero inteiro entre -128 e 127 (1 byte)
# Obs: Se a coluna tiver valores maiores que 127 ou menores que -128, vai dar erro.

df_opt['Idade'] = df_opt['Idade'].astype(np.int8)  # exemplo: se Idade varia entre 0 e 100
df_opt['Pontuacao_Gasto'] = df_opt['Pontuacao_Gasto'].astype(np.int8)  # idem para Score/Pontua√ß√£o
df_opt['Anos_Cliente'] = df_opt['Anos_Cliente'].astype(np.int8)  # idem para tempo de cliente

# Alternativas:
# np.int16: -32.768 a 32.767 (2 bytes)
# np.int32: -2 bi a 2 bi (4 bytes)
# np.int64: padr√£o do Pandas (8 bytes), s√≥ use se necess√°rio


# === CONVERS√ÉO DE STRINGS/CATEGORIAS PARA TYPE 'category' ===

# Colunas de texto com **baixa cardinalidade** (poucos valores √∫nicos) s√£o candidatas ideais para o tipo 'category'.
# Isso reduz drasticamente o uso de mem√≥ria e acelera opera√ß√µes como groupby, merge e filtragem.
# Exemplo: G√™nero = ['Masculino', 'Feminino', 'Outro'] ‚Üí ideal p/ category

for col in ['Genero', 'Regiao', 'Categoria_Pref', 'Faixa_Etaria']:
    if col in df_opt.columns:  # verifica se a coluna existe no DataFrame
        df_opt[col] = df_opt[col].astype('category')  # converte para tipo categ√≥rico

# Alternativas:
# - 'category' com par√¢metro ordered=True para ordens espec√≠ficas (ex: Baixo < M√©dio < Alto)
# - Cuidado ao usar com muitas categorias √∫nicas (alta cardinalidade), pois pode piorar a performance


# === RESULTADOS AP√ìS A OTIMIZA√á√ÉO ===

print("\nDtypes Otimizados:")
print(df_opt.dtypes)

# Recalcula e imprime o novo uso de mem√≥ria ap√≥s otimizar os tipos.
print(f"Uso de mem√≥ria otimizado: {df_opt.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
````

* **Opera√ß√µes Vetorizadas:** Evite iterar sobre linhas (`.iterrows()`, `.apply(axis=1)` √© lento). Use opera√ß√µes NumPy/Pandas que atuam no array/Series inteiro.
* **Indexa√ß√£o:** Use indexa√ß√£o eficiente (`.loc`, `.iloc`, booleana). √çndices ordenados (`.sort_index()`) podem acelerar sele√ß√µes baseadas em r√≥tulos.
* **Bibliotecas Externas:** Para dados maiores que a mem√≥ria, considere Dask, Vaex ou processamento distribu√≠do (Spark com PySpark). Para acelerar c√≥digo Python num√©rico, considere Numba ou Cython.
