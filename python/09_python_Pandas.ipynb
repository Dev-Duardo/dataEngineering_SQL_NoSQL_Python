**1. Introdução ao Pandas**

**1.1. O que é Pandas?**

Pandas é uma biblioteca de software fundamental para a linguagem de programação Python, projetada especificamente para **manipulação e análise de dados**. Lançada como código aberto sob a licença BSD ⁷, ela se estabeleceu como uma ferramenta essencial no ecossistema de ciência de dados em Python. O nome "Pandas" deriva do termo "Panel Data" (Dados em Painel), um conceito econométrico para conjuntos de dados multidimensionais, e também de "Python Data Analysis".³, ⁸

A biblioteca se destaca por oferecer estruturas de dados de alto desempenho e fáceis de usar, juntamente com um conjunto rico de ferramentas para análise de dados. Seu foco principal reside em trabalhar com dados estruturados, que podem ser tabulares (como planilhas ou tabelas SQL), multidimensionais ou heterogêneos.⁸ Pandas visa ser a ferramenta de análise e manipulação de dados de código aberto mais poderosa e flexível disponível.⁹

As características chave do Pandas incluem sua rapidez, poder de processamento, flexibilidade e facilidade de uso.⁸ Ele fornece duas estruturas de dados primárias:

* **Series:** Um array unidimensional rotulado, capaz de armazenar dados de qualquer tipo (inteiros, strings, floats, objetos Python, etc.). Possui um **índice** que rotula cada elemento.⁸
* **DataFrame:** Uma estrutura de dados tabular bidimensional, com eixos rotulados (linhas e colunas). Colunas podem ter tipos diferentes. Visualizável como uma planilha ou tabela SQL.⁸

**1.2. Importância na Análise de Dados e Ciência de Dados**

Pandas é frequentemente considerada a biblioteca *mais crucial* para análise de dados e ciência de dados em Python.⁹ Sua relevância deriva de:

* **Manipulação Eficiente de Dados Estruturados:** Facilita enormemente o trabalho com dados tabulares.⁸
* **Conjunto Abrangente de Ferramentas:** Cobre todo o ciclo de análise: leitura/escrita, limpeza, transformação, fusão, tratamento de dados ausentes, seleção, filtragem, agregação, remodelação e visualização.⁸
* **Integração com o Ecossistema Python:** Integra-se perfeitamente com NumPy (sua base), Matplotlib, Seaborn (visualização) e Scikit-learn (Machine Learning).⁸
* **Análise Exploratória de Dados (EDA):** Essencial para calcular estatísticas, identificar padrões, visualizar distribuições e entender os dados.⁸
* **Preparação para Machine Learning:** Indispensável para engenharia de features, codificação de variáveis, normalização, etc.⁹

Pandas atua como a ponte entre dados brutos e análises avançadas ou modelos de ML, simplificando tarefas complexas.⁹

**1.3. Estruturas de Dados Fundamentais (Visão Geral)**

* **Series:** Uma coluna. Array 1D NumPy com um índice (rótulos).
* **DataFrame:** Uma tabela. Múltiplas colunas (Series) compartilhando um índice de linha.

Construídas sobre NumPy para eficiência, mas com rótulos (índices) para acesso e manipulação flexível e intuitiva, ideal para dados do mundo real.¹⁰

**3. Estruturas de Dados Pandas**

**3.1. Series**

Objeto 1D tipo array com um índice (`index`). Armazena dados homogêneos (mas pode cair para `dtype=object`). Tamanho imutável, valores mutáveis.⁸, ⁹, ¹⁰

* **Criação (`pd.Series()`):**
    * A partir de lista/array NumPy (índice padrão ou customizado).¹⁰
    * A partir de dicionário (chaves viram índice).¹⁰
    * A partir de escalar (requer índice).²⁵

````
print("\n--- 3.1 Series ---")

# --------------------------------------
# 1️⃣ Criando uma Series a partir de uma lista, com índice customizado
# --------------------------------------
dados_lista = [10, 20, 30]  # Lista de valores que vão compor os dados da Series
indices = ['a', 'b', 'c']   # Índices personalizados para cada valor da lista

# pd.Series(data, index, name) -> Cria uma Series com os dados e índices definidos
# - 'data' pode ser lista, array, dicionário, escalar...
# - 'index' é opcional: se não for passado, será um RangeIndex padrão (0, 1, 2, ...)
# - 'name' é o nome da Series, útil para identificação ou exportação
s_lista = pd.Series(data=dados_lista, index=indices, name='MinhaSerie')

print("Series de lista com índice e nome:")
print(s_lista)

# Resultado esperado:
# a    10
# b    20
# c    30
# Name: MinhaSerie, dtype: int64

# --------------------------------------
# 2️⃣ Criando uma Series a partir de um dicionário
# --------------------------------------
dados_dict = {'x': 100, 'y': 200, 'z': 300}
# Quando o argumento 'data' é um dicionário:
# - As chaves se tornam os índices
# - Os valores se tornam os dados da Series
s_dict = pd.Series(dados_dict)

print("\nSeries de dicionário:")
print(s_dict)

# Resultado esperado:
# x    100
# y    200
# z    300
# dtype: int64

# --------------------------------------
# 3️⃣ Criando uma Series a partir de um escalar
# --------------------------------------
# Um escalar é um único valor (ex: 5)
# Quando passamos um escalar como 'data' e uma lista como 'index':
# - O valor escalar é repetido automaticamente para cada índice
s_esc = pd.Series(5, index=['p', 'q', 'r'])

print("\nSeries de escalar:")
print(s_esc)

# Resultado esperado:
# p    5
# q    5
# r    5
# dtype: int64
````

````
import pandas as pd

# Vamos reutilizar a Series de exemplo
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'], name='MinhaSerie')

# 🔍 .values → Retorna os dados como um array NumPy
print("s.values →", s.values)
# Resultado: array([10, 20, 30])
# Obs: útil para operações vetorizadas com NumPy

# 🔍 .index → Retorna o objeto Index (os rótulos/índices)
print("s.index →", s.index)
# Resultado: Index(['a', 'b', 'c'], dtype='object')
# Obs: isso é um objeto especial do Pandas, com métodos próprios (ex: .tolist())

# 🔍 .dtype → Retorna o tipo dos dados armazenados
print("s.dtype →", s.dtype)
# Resultado: int64
# Obs: pode ser int, float, object (texto), category, datetime, etc.

# 🔍 .name → Retorna (ou define) o nome da Series
print("s.name →", s.name)
# Resultado: MinhaSerie
# Obs: útil ao exportar ou ao usar como coluna em DataFrame

# 🔍 .size → Número total de elementos
print("s.size →", s.size)
# Resultado: 3
# Obs: igual a len(s), ou seja, quantidade de elementos

# 🔍 .shape → Formato (tupla), igual ao array NumPy
print("s.shape →", s.shape)
# Resultado: (3,)
# Obs: o número de linhas. Como Series é 1D, vem como tupla de um valor
````

🧠 Dica Rápida:
Esses atributos não são métodos (não têm `()`) — eles são como "etiquetas" de metadados embutidas no objeto `Series`. São muito úteis pra diagnósticos e para entender rapidamente o que tem ali dentro sem precisar printar tudo.

````
import pandas as pd

print("\nAcesso e Operações em Series:")

# Series de exemplo
s_lista = pd.Series(data=[10, 20, 30], index=['a', 'b', 'c'], name='MinhaSerie')

# 🔍 Acesso por rótulo
print(f"Acesso por rótulo 'b': {s_lista['b']}")
# → Retorna o valor associado ao índice 'b' (20).
# → Isso é o modo padrão e intuitivo de acessar elementos em uma Series.
# → É como acessar um valor em um dicionário Python.

# 🔍 Acesso por posição (index inteiro)
print(f"Acesso por posição 1: {s_lista.iloc[1]}")
# → Usa o método `.iloc[]` para acessar pela posição (índice numérico zero-based).
# → Aqui, posição 1 corresponde ao segundo elemento (índice 'b').

# 🔍 Slicing por rótulo
print(f"Slice por rótulo 'a':'b':\n{s_lista['a':'b']}")
# → Retorna os elementos de 'a' até 'b' (inclusive 'b'!).
# → Importante: no slicing por rótulo, o limite superior é INCLUSIVO.

# 🔍 Slicing por posição
print(f"Slice por posição 0:2:\n{s_lista.iloc[0:2]}")
# → Retorna os elementos da posição 0 até 1 (exclui a posição 2).
# → Como listas Python, o limite superior é EXCLUSIVO.

# 🔍 Indexação Booleana
print(f"Filtro (valor > 15):\n{s_lista[s_lista > 15]}")
# → Retorna apenas os valores maiores que 15.
# → Isso é super útil em filtros, análises, e preparação de dados.

# 🔍 Operação vetorizada
print(f"Operação vetorizada (s_lista * 2):\n{s_lista * 2}")
# → Multiplica cada valor da Series por 2.
# → Operações vetorizadas são eficientes, rápidas e seguras.
# → Pandas alinha automaticamente os índices ao fazer operações entre Series.

# 🔄 Conversão de tipo com .astype()
s_float = s_lista.astype(float)
print(f"\nSeries convertida para float:\n{s_float}")
# → Converte os valores da Series de int para float (10 → 10.0).
# → .astype() é muito usado para otimização, compatibilidade ou operações específicas.
# → Pode converter para qualquer tipo compatível: int, float, str, category, etc.
````

````
import pandas as pd
import numpy as np

print("\n--- 3.2 DataFrame ---")

# 🧱 1. Criando um DataFrame a partir de um dicionário de listas:
dados_pessoas = {
    'Nome': ['Alice', 'Bob', 'Charlie'],  # Lista de nomes (coluna 1)
    'Idade': [25, 30, 22],                # Lista de idades (coluna 2)
    'Cidade': ['A', 'B', 'A']             # Lista de cidades (coluna 3)
}
df_pessoas = pd.DataFrame(dados_pessoas)
print("DataFrame de dict de listas:")
print(df_pessoas)

# ✅ Esse método é o mais direto: as **chaves do dicionário viram colunas** e os **valores (listas)** são os dados.
# ✅ As listas precisam ter o **mesmo comprimento**, senão o pandas vai reclamar.

# 🧱 2. Criando um DataFrame a partir de uma lista de dicionários:
dados_registros = [
    {'id': 1, 'valor': 100, 'status': 'ok'},
    {'id': 2, 'valor': 150},  # 'status' ausente aqui — pandas preenche com NaN (Not a Number)
    {'id': 3, 'valor': 120, 'status': 'pendente'}
]
df_registros = pd.DataFrame(dados_registros)
print("\nDataFrame de lista de dicts:")
print(df_registros)

# ✅ Cada dicionário representa uma **linha** no DataFrame.
# ✅ As **chaves se tornam os nomes das colunas**.
# ✅ Se alguma chave estiver ausente em um dicionário, pandas preenche com NaN (valor ausente/nulo).

# 🧱 3. Criando um DataFrame a partir de um array NumPy 2D:
dados_np = np.random.randn(3, 4)  # Gera uma matriz 3x4 com valores aleatórios da distribuição normal padrão
indice_linhas = pd.date_range('2024-01-01', periods=3, freq='D')  # Índices (linhas) como datas sequenciais
nomes_colunas = ['Sensor1', 'Sensor2', 'Sensor3', 'Sensor4']       # Nomes para as colunas

df_sensores = pd.DataFrame(dados_np, index=indice_linhas, columns=nomes_colunas)
print("\nDataFrame de array NumPy:")
print(df_sensores.round(2))  # Arredondando para 2 casas decimais só pra ficar mais bonito

# ✅ Esse método é útil quando você já tem os dados em NumPy e quer organizar em tabela.
# ✅ Você pode passar `index` e `columns` para deixar a tabela legível e informativa.

# --------------------------------------
# 📌 Recapitulando os modos de criação:
# pd.DataFrame(dicionário de listas)     -> Cada chave vira uma coluna
# pd.DataFrame(lista de dicionários)     -> Cada dict é uma linha, colunas são unificadas
# pd.DataFrame(array 2D, columns, index) -> Constrói a tabela com base em dados já existentes (como NumPy)

# --------------------------------------
# 🧠 Curiosidade Técnica:
# DataFrame = conjunto de Series alinhadas pelo mesmo índice
# Ou seja, cada coluna é uma `Series`, e o DataFrame é um "dicionário" de Series.
````

````
# Vamos explorar os principais atributos de um DataFrame — aqui usamos `df_sensores`
# Esse DF foi criado com dados aleatórios de sensores, datas como índice e 4 colunas numéricas.

print(f"\nAtributos de df_sensores:")

# 🔹 .shape: mostra a forma da tabela -> (número de linhas, número de colunas)
print(f"  Shape: {df_sensores.shape}")  # Exemplo: (3, 4)

# 🔹 .dtypes: tipo de dado de cada coluna
print(f"  Dtypes:\n{df_sensores.dtypes}")  # Mostra se é float, int, object, etc.

# 🔹 .columns: retorna um objeto Index com os nomes das colunas
print(f"  Columns: {df_sensores.columns}")  # Ex: Index(['Sensor1', 'Sensor2', ...])

# 🔹 .index: retorna o índice (linhas), que neste caso são datas
print(f"  Index: {df_sensores.index}")  # Ex: DatetimeIndex([...], freq='D')

# 🔹 .info(): fornece um resumo geral do DataFrame
# Inclui: total de entradas (linhas), colunas, dtypes, quantos não-nulos em cada coluna e uso de memória
print("\n  Info:")
df_sensores.info()

# 🔹 .describe(): estatísticas descritivas só para colunas numéricas
# Inclui: count, mean, std, min, 25%, 50%, 75%, max
print("\n  Describe:")
print(df_sensores.describe().round(2))  # Arredondado pra facilitar a leitura

# 🔹 .values: retorna os dados como array NumPy 2D
# ⚠️ Útil pra interoperar com bibliotecas que usam NumPy, mas perde rótulos de linha/coluna
print(f"\n  Valores (.values):\n{df_sensores.values.round(2)}")  # Os dados crus como matriz

# 🔹 .T: transposição — troca linhas por colunas e vice-versa
# Pode ser útil em análise exploratória e visualização
print(f"\n  Transposto (.T):\n{df_sensores.T.round(2)}")
````

````
# -*- coding: utf-8 -*-
"""
Este bloco de código demonstra e explica em detalhes como usar a função
pd.read_csv do Pandas para ler dados de arquivos CSV (ou similares),
lidando com diversas complexidades comuns encontradas nesses arquivos.
Inclui explicações detalhadas sobre a sintaxe e os parâmetros utilizados.
"""

# Importa a classe StringIO do módulo io.
# StringIO permite tratar uma string em memória como se fosse um arquivo,
# o que é útil para testar ou quando os dados CSV estão em uma variável string.
from io import StringIO

# Importa a biblioteca Pandas. Assume-se que já foi importada,
# mas é uma boa prática garantir. 'pd' é o alias padrão.
import pandas as pd

# ==============================================================================
# 4. Leitura e Escrita de Dados (Contexto Fornecido)
# ==============================================================================

# Comentários iniciais fornecidos pelo usuário (mantidos para contexto):
# *(Tabela de Resumo dos Atributos Fundamentais omitida para brevidade, já presente na versão anterior)*
#
# **4. Leitura e Escrita de Dados**
#
# Pandas IO Tools: Funções pd.read_* e métodos df.to_* para diversos formatos.³⁶
#
# **4.1. Lendo Dados**
#
# * **pd.read_csv()**: Para CSV e arquivos delimitados.³⁶
#     * Parâmetros detalhados: filepath_or_buffer, sep, header, names, index_col,
#       usecols, dtype, encoding, skiprows, nrows, na_values, parse_dates,
#       skipfooter, comment, thousands, decimal, chunksize (para ler em pedaços).⁶, ³⁷, ³⁸

print("\n--- 4.1 Lendo Dados com pd.read_csv ---")

# ==============================================================================
# Simulação de Dados CSV Complexos
# ==============================================================================

# Cria uma string multi-linhas que simula o conteúdo de um arquivo CSV.
# Este exemplo inclui várias características que exigem parâmetros específicos
# na função pd.read_csv para serem lidos corretamente:
# 1. Linha de comentário inicial (#).
# 2. Separador de colunas não padrão (';').
# 3. Cabeçalho na segunda linha (após o comentário).
# 4. Números formatados com separador de milhar ('.') e decimal (',').
# 5. Um valor ausente representado por um caractere específico ('?').
# 6. Uma linha de rodapé no final que deve ser ignorada.
# 7. Datas em formato de string que precisam ser convertidas.
csv_complexo = """# Comentário a ser ignorado
ID;Data;Valor;Cidade
1;2024-01-10;1.250,50;Rio de Janeiro
2;2024-01-11;850,00;São Paulo
3;2024-01-12;?;Salvador
Rodapé a ser ignorado"""
# Nota: A quebra de linha após a última linha de dados e antes do rodapé é importante
# para que 'skipfooter' funcione como esperado na string simulada.

# ==============================================================================
# Leitura do CSV Simulado com Parâmetros Detalhados
# ==============================================================================

# Função: pd.read_csv(filepath_or_buffer, sep=',', delimiter=None, header='infer', names=None, index_col=None, usecols=None, dtype=None, engine=None, converters=None, true_values=None, false_values=None, skipinitialspace=False, skiprows=None, skipfooter=0, nrows=None, na_values=None, keep_default_na=True, na_filter=True, verbose=False, skip_blank_lines=True, parse_dates=False, infer_datetime_format=False, keep_date_col=False, date_parser=None, dayfirst=False, cache_dates=True, iterator=False, chunksize=None, compression='infer', thousands=None, decimal='.', lineterminator=None, quotechar='"', quoting=0, doublequote=True, escapechar=None, comment=None, encoding=None, dialect=None, error_bad_lines=None, warn_bad_lines=None, on_bad_lines=None, delim_whitespace=False, low_memory=True, memory_map=False, float_precision=None, storage_options=None)
# Objetivo Principal: Ler um arquivo CSV (ou similar) e carregá-lo em um DataFrame do Pandas.
# Retorna: Um DataFrame do Pandas (ou um iterador se chunksize for especificado).

try:
    # Tentativa de ler os dados usando pd.read_csv com parâmetros ajustados
    df_csv_complexo = pd.read_csv(
        # --- Parâmetro 1: filepath_or_buffer ---
        # Fonte dos dados. Pode ser um caminho de arquivo (string), URL (string)
        # ou um objeto 'file-like' (com um método read()).
        # Aqui usamos StringIO para tratar nossa string 'csv_complexo' como um arquivo.
        StringIO(csv_complexo),

        # --- Parâmetro 2: sep (ou delimiter) ---
        # O caractere (ou expressão regular) que separa as colunas no arquivo.
        # Valor: ';' -> Define o ponto e vírgula como separador. O padrão é ','.
        sep=';',

        # --- Parâmetro 3: skiprows ---
        # Número de linhas a pular do *início* do arquivo antes de começar a ler os dados.
        # Pode ser um inteiro N (pula as primeiras N linhas) ou uma lista de índices de linha a pular.
        # Valor: 1 -> Pula a primeira linha (índice 0), que é a linha de comentário.
        # Nota: Após pular linhas, o Pandas procura o cabeçalho na próxima linha,
        #       a menos que o parâmetro 'header' especifique outra linha ou seja None.
        #       Neste caso, após pular 1 linha, a linha seguinte ("ID;Data;Valor;Cidade")
        #       será corretamente identificada como cabeçalho (comportamento padrão header=0 relativo ao início da leitura).
        skiprows=1,

        # --- Parâmetro 4: comment ---
        # Caractere que, se aparecer no início de uma linha (após espaços em branco),
        # indica que o restante dessa linha é um comentário e deve ser ignorado.
        # Valor: '#' -> Linhas começando com '#' serão ignoradas.
        #        Embora a primeira linha já tenha sido pulada por `skiprows`,
        #        este parâmetro protegeria contra comentários em outras partes do arquivo.
        comment='#',

        # --- Parâmetro 5: skipfooter ---
        # Número de linhas a ignorar no *final* do arquivo.
        # Valor: 1 -> Ignora a última linha ("Rodapé a ser ignorado").
        # Importante: Este parâmetro só funciona se o 'engine' for 'python'.
        skipfooter=1,

        # --- Parâmetro 6: engine ---
        # Motor de análise (parser engine) a ser usado. Opções: 'c', 'python', 'pyarrow'.
        # Valor: 'python' -> Necessário para usar 'skipfooter'. O motor 'python' é mais
        #          flexível, mas geralmente mais lento que o motor 'c' (padrão).
        #          O motor 'c' é otimizado para velocidade, mas tem menos recursos.
        #          'pyarrow' é uma opção mais recente e rápida para certos cenários.
        engine='python',

        # --- Parâmetro 7: thousands ---
        # Caractere a ser interpretado como separador de milhares ao converter strings em números.
        # Valor: '.' -> Reconhece o ponto como separador de milhar (ex: "1.250").
        thousands='.',

        # --- Parâmetro 8: decimal ---
        # Caractere a ser interpretado como separador decimal ao converter strings em números.
        # Valor: ',' -> Reconhece a vírgula como separador decimal (ex: "250,50").
        # Nota: 'thousands' e 'decimal' são essenciais para ler corretamente números
        #       em formatos não padrão (ex: formato monetário brasileiro ou europeu).
        decimal=',',

        # --- Parâmetro 9: na_values ---
        # String ou lista de strings adicionais que devem ser reconhecidas como NaN (Not a Number / Ausente).
        # Valor: ['?'] -> A string exata "?" será interpretada como um valor ausente (NaN).
        #        Por padrão, o Pandas já reconhece strings como '', '#N/A', 'NaN', 'NULL', etc.
        #        Use `keep_default_na=False` se quiser *apenas* os valores em `na_values`.
        na_values=['?'],

        # --- Parâmetro 10: parse_dates ---
        # Tenta converter colunas especificadas para o tipo datetime do Pandas.
        # Valor: ['Data'] -> Instrui o Pandas a tentar analisar a coluna chamada 'Data' como data/hora.
        #        Pode ser True (tenta analisar o índice), uma lista de nomes/índices de colunas,
        #        ou um dicionário para combinar múltiplas colunas em uma data.
        #        Para performance em arquivos grandes, considere converter após a leitura com
        #        `pd.to_datetime(df['Data'], format='%Y-%m-%d')` se o formato for conhecido.
        parse_dates=['Data']

        # (Outros parâmetros comuns e úteis não usados aqui incluem:
        #  header=None -> Se não houver linha de cabeçalho.
        #  names=['A','B','C'] -> Fornece nomes de colunas (útil com header=None).
        #  index_col='ID' -> Usa a coluna 'ID' como índice do DataFrame.
        #  usecols=['ID','Valor'] -> Lê apenas um subconjunto de colunas (economiza memória).
        #  dtype={'ID':int, 'Valor':float} -> Especifica o tipo de dados para colunas (economia/precisão).
        #  encoding='latin-1' -> Especifica a codificação do arquivo (importante para acentos, etc.).
        #  nrows=100 -> Lê apenas as N primeiras linhas de dados.
        # )
    )

    # Imprime o DataFrame resultante da leitura
    print("\nLeitura CSV Complexo (DataFrame resultante):")
    print(df_csv_complexo)

    # Imprime informações sobre o DataFrame, incluindo tipos de dados por coluna,
    # contagem de valores não nulos e uso de memória.
    # Útil para verificar se a leitura (tipos, NaNs, datas) funcionou como esperado.
    print("\nInformações do DataFrame (.info()):")
    df_csv_complexo.info()

except Exception as e:
    # Captura e informa qualquer erro que possa ocorrer durante a leitura
    print(f"\nErro ao ler o CSV simulado: {e}")
    # Um problema comum poderia ser um FileNotFoundError se estivesse lendo um arquivo real,
    # ou um erro de parsing se os parâmetros não correspondessem ao formato dos dados.
    # Neste caso, como usamos StringIO, erros seriam mais prováveis por parâmetros incorretos.

print("-" * 50) # Separador visual

# ==============================================================================
# Leitura em Chunks (Conceito para Arquivos Grandes)
# ==============================================================================

# --- Parâmetro: chunksize ---
# Se você fornecer um valor inteiro para `chunksize=N`, a função `pd.read_csv`
# não retorna um DataFrame completo, mas sim um *iterador*.
# Cada vez que você itera sobre ele, ele lê e retorna as próximas N linhas
# do arquivo como um DataFrame (o "chunk").
# Objetivo: Processar arquivos que são muito grandes para serem carregados
#           inteiramente na memória RAM. Permite processar o arquivo em pedaços.

print("\nLeitura CSV em Chunks (exemplo conceitual - código comentado):")

# O código abaixo está comentado porque requer um arquivo real chamado 'arquivo_grande.csv'.
# Ele demonstra a sintaxe e o loop para processar chunks.
"""
try:
    # Define o tamanho do chunk (quantas linhas ler por vez)
    tamanho_chunk = 10000

    # Cria o iterador. Nenhum dado é lido neste ponto ainda.
    chunk_iter = pd.read_csv('arquivo_grande.csv', chunksize=tamanho_chunk)
                             # Pode incluir outros parâmetros como sep, header, etc.

    print(f"Iniciando leitura de 'arquivo_grande.csv' em chunks de {tamanho_chunk} linhas...")

    # Lista para armazenar resultados processados de cada chunk (exemplo)
    resultados_processados = []

    # Loop sobre o iterador. A cada iteração, lê o próximo chunk.
    for i, chunk_df in enumerate(chunk_iter):
        # chunk_df é um DataFrame contendo as linhas do chunk atual
        print(f"Processando chunk {i+1} com shape {chunk_df.shape}...")

        # --- Realize aqui o processamento desejado para cada chunk ---
        # Exemplo: Filtrar dados, calcular estatísticas, transformar colunas, etc.
        # media_valor_chunk = chunk_df['Valor'].mean()
        # resultados_processados.append(media_valor_chunk)
        # -------------------------------------------------------------

        # Libera a memória do chunk atual explicitamente se necessário (opcional)
        # del chunk_df

    print("Processamento de todos os chunks concluído.")
    # Se agregou resultados, pode combiná-los agora:
    # media_total = pd.Series(resultados_processados).mean()
    # print(f"Resultado agregado (exemplo: média total): {media_total}")

except FileNotFoundError:
    print("Erro: O arquivo 'arquivo_grande.csv' não foi encontrado para demonstrar chunking.")
except Exception as e:
    print(f"Erro durante o processamento em chunks: {e}")
"""
# Mensagem indicando que o código acima é um exemplo conceitual.
print("(O código de exemplo para chunking está comentado pois requer um arquivo real)")
print("-" * 50)

print("\n--- Fim da Seção 4.1: Leitura de Dados com pd.read_csv ---")

# --- Resumo e Pontos Chave ---
# - `pd.read_csv` é extremamente flexível devido aos seus numerosos parâmetros.
# - Entender parâmetros como `sep`, `header`, `skiprows`, `skipfooter`, `comment`, `decimal`, `thousands`, `na_values`, `parse_dates`, e `encoding` é crucial para ler dados "reais" corretamente.
# - O parâmetro `engine` ('c' vs 'python') afeta performance e disponibilidade de recursos (como `skipfooter`).
# - `StringIO` é útil para testar `read_csv` com dados em strings.
# - `chunksize` é a solução para ler arquivos que excedem a memória disponível.
# - Sempre verifique o DataFrame resultante com `.head()`, `.info()`, `.describe()` para garantir que a leitura ocorreu como esperado.
````

* **`pd.read_sql...()`**: Para bancos de dados SQL.³⁶ Requer conector + SQLAlchemy (recomendado).⁴⁴
    * Parâmetros detalhados: `sql` (query ou nome tabela), `con` (conexão), `index_col`, `params` (prevenir SQL Injection), `parse_dates`, `chunksize`.⁴⁴

````
# -*- coding: utf-8 -*-
"""
Este bloco de código demonstra como interagir com um banco de dados SQLite
usando a biblioteca `sqlite3` (padrão do Python) e como ler os resultados
de uma consulta SQL diretamente para um DataFrame Pandas usando `pd.read_sql_query`.
Inclui explicações detalhadas sobre as funções, sintaxe e parâmetros.
"""

# Importa a biblioteca sqlite3, que fornece a interface para bancos de dados SQLite.
# SQLite é um banco de dados leve, baseado em arquivo, muito comum e que já vem com Python.
import sqlite3

# Importa a biblioteca Pandas (assume-se que já foi importada anteriormente).
# 'pd' é o alias convencional.
import pandas as pd

print("\n--- 4.x Leitura de Dados de Banco SQL (SQLite) ---")
# (A numeração 4.x é genérica, ajuste conforme a estrutura do seu documento)

# ==============================================================================
# Conexão e Preparação do Banco de Dados SQLite (Exemplo em Memória)
# ==============================================================================

# Tentaremos executar as operações de banco de dados dentro de um bloco try...finally
# para garantir que a conexão seja fechada mesmo se ocorrer um erro.
conn = None # Inicializa a variável de conexão fora do try
try:
    # --- Estabelecendo a Conexão ---
    # Função: sqlite3.connect(database_path)
    # Objetivo: Criar ou conectar a um banco de dados SQLite.
    # Parâmetros:
    #   database_path (str): O caminho para o arquivo do banco. Se não existir, será criado.
    #       Valor ':memory:': Cria um banco de dados volátil *apenas na memória RAM*.
    #           Ele é muito rápido e útil para exemplos, testes ou dados temporários,
    #           pois desaparece completamente quando a conexão é fechada.
    # Retorna: Um objeto Connection que representa a conexão com o banco.
    conn = sqlite3.connect(':memory:')
    print("Conexão com banco de dados SQLite em memória estabelecida.")

    # --- Executando Comandos SQL (DDL e DML) ---
    # Método: connection.execute(sql_statement, parameters=None)
    # Objetivo: Executa uma única instrução SQL no banco de dados.
    #          É um método conveniente que cria um cursor, executa e retorna o cursor.
    # Parâmetros:
    #   sql_statement (str): A instrução SQL a ser executada.
    #   parameters (tuple/dict, opcional): Usado para passar valores de forma segura
    #             em instruções que usam placeholders (como SELECT, INSERT, UPDATE).
    #             Não tipicamente usado para DDL (CREATE, ALTER, DROP).

    # 1. Criar Tabela (DDL - Data Definition Language)
    #    Define a estrutura da tabela 'funcionarios'.
    #    Tipos de dados SQLite comuns: INTEGER, REAL (ponto flutuante), TEXT, BLOB, NULL.
    create_table_sql = "CREATE TABLE funcionarios (id INTEGER PRIMARY KEY, nome TEXT, salario REAL)"
    print(f"Executando DDL: {create_table_sql}")
    conn.execute(create_table_sql)

    # 2. Inserir Dados (DML - Data Manipulation Language)
    #    Adiciona duas linhas à tabela 'funcionarios'.
    insert_sql_1 = "INSERT INTO funcionarios (id, nome, salario) VALUES (1, 'Alice', 5000.0)"
    insert_sql_2 = "INSERT INTO funcionarios (id, nome, salario) VALUES (?, ?, ?)" # Usando placeholders
    params_insert_2 = (2, 'Bob', 6500.5)
    print(f"Executando DML (INSERT 1): {insert_sql_1}")
    conn.execute(insert_sql_1)
    print(f"Executando DML (INSERT 2): {insert_sql_2} com params {params_insert_2}")
    conn.execute(insert_sql_2, params_insert_2) # Passando parâmetros de forma segura

    # --- Commit das Alterações ---
    # Método: connection.commit()
    # Objetivo: Salva permanentemente todas as alterações feitas desde o último commit
    #          (ou desde o início da conexão se for o primeiro). Essencial para que
    #          CREATE TABLE e INSERT/UPDATE/DELETE tenham efeito duradouro no banco.
    conn.commit()
    print("Alterações commitadas no banco de dados.")

    # ==============================================================================
    # Leitura de Dados do Banco com Pandas
    # ==============================================================================
    print("\nLeitura de Banco SQL (SQLite) usando pd.read_sql_query:")

    # --- Definindo a Query SQL e Parâmetros ---
    # Query para selecionar funcionários com salário acima de um certo valor.
    # O '?' é um placeholder (estilo 'qmark' padrão do sqlite3) que será
    # substituído de forma segura pelo valor fornecido no parâmetro `params`.
    query = "SELECT * FROM funcionarios WHERE salario > ?"
    print(f"Query SQL a ser executada: {query}")

    # Parâmetros para a query. Deve ser uma sequência (tupla ou lista) com
    # tantos elementos quantos forem os placeholders '?'.
    # A vírgula após 6000 é crucial para criar uma tupla de um elemento.
    # *** Segurança: Usar `params` é a forma correta e segura de passar valores
    # para queries SQL, prevenindo ataques de SQL Injection. NUNCA insira
    # valores diretamente na string da query usando formatação de string (f-string, %)
    # se esses valores vierem de fontes externas ou não confiáveis. ***
    params_sql = (6000,)
    print(f"Parâmetros para a query: {params_sql}")

    # --- Usando pd.read_sql_query ---
    # Função: pd.read_sql_query(sql, con, index_col=None, coerce_float=True, params=None, parse_dates=None, chunksize=None)
    # Objetivo: Executar uma consulta SQL e ler os resultados diretamente para um DataFrame Pandas.
    # Parâmetros Chave:
    #   sql (str): A string contendo a consulta SQL (pode ter placeholders).
    #   con (object): O objeto de conexão com o banco de dados (DBAPI2 connection - como `conn` do sqlite3)
    #                 ou uma engine/string de conexão SQLAlchemy.
    #   params (list, tuple ou dict, opcional): Sequência ou dicionário de parâmetros
    #              para passar à consulta SQL de forma segura, substituindo os placeholders.
    #   index_col (str ou list, opcional): Coluna(s) do resultado SQL a ser(em) usada(s) como índice do DataFrame.
    #   parse_dates (list ou dict, opcional): Coluna(s) a serem interpretadas como datas.
    # Retorna: Um DataFrame Pandas com os resultados da consulta.

    df_sqlite = pd.read_sql_query(query, conn, params=params_sql)

    # Imprime o DataFrame resultante
    print("\nDataFrame resultante da query:")
    print(df_sqlite)
    print("\nTipos de dados do DataFrame:")
    df_sqlite.info() # Mostra que Pandas inferiu os tipos corretamente do SQL

except sqlite3.Error as e:
    # Captura erros específicos do sqlite3 (ex: erro de sintaxe SQL)
    print(f"\nErro SQLite ocorrido: {e}")
except Exception as e:
    # Captura outros erros genéricos
    print(f"\nUm erro inesperado ocorreu: {e}")

finally:
    # --- Fechando a Conexão ---
    # O bloco `finally` garante que este código seja executado SEMPRE,
    # quer o bloco `try` tenha sido bem-sucedido ou tenha gerado um erro.
    if conn: # Verifica se a conexão foi estabelecida com sucesso antes de tentar fechar
        # Método: connection.close()
        # Objetivo: Fecha a conexão com o banco de dados, liberando recursos.
        #          Para bancos :memory:, isso destrói o banco.
        #          Para bancos em arquivo, garante que tudo foi escrito e desbloqueia o arquivo.
        conn.close()
        print("\nConexão com o banco de dados fechada no bloco finally.")

print("-" * 50)

# ==============================================================================
# Observação sobre Gerenciador de Contexto (`with`) - Método Preferido
# ==============================================================================
# A forma mais idiomática e segura de trabalhar com conexões `sqlite3` (e muitos
# outros recursos que precisam ser fechados) em Python é usando um gerenciador
# de contexto (`with` statement). Ele lida automaticamente com o commit/rollback
# e o fechamento da conexão.

print("\nExemplo usando 'with' (método recomendado):")
try:
    # O 'with' garante que conn_with será fechado no final, e lida com commits/rollbacks.
    with sqlite3.connect(':memory:') as conn_with:
        conn_with.execute("CREATE TABLE exemplo_with (id INT, nome TEXT)")
        conn_with.execute("INSERT INTO exemplo_with VALUES (?, ?)", (10, 'Com With'))
        # Commit é geralmente automático ao sair do 'with' sem erro.
        # Rollback é automático se ocorrer um erro dentro do 'with'.

        df_with = pd.read_sql_query("SELECT * FROM exemplo_with", conn_with)
        print("DataFrame lido dentro do 'with':")
        print(df_with)
    # A conexão conn_with já está fechada aqui.
    print("Bloco 'with' concluído, conexão fechada automaticamente.")
except Exception as e:
    print(f"Erro no exemplo com 'with': {e}")
    # A conexão ainda seria fechada automaticamente.

print("-" * 50)
print("--- Fim da Seção de Leitura SQL com SQLite ---")
````

* **.to_sql()**: Salva DataFrame em tabela SQL.³⁶ Requer SQLAlchemy.⁵³, ⁵⁴
    * Parâmetros detalhados: `name` (tabela), `con` (conexão), `schema`, `if_exists` ('fail', 'replace', 'append'), `index`, `index_label`, `chunksize` (para inserção em lotes), `dtype` (dicionário de tipos SQLAlchemy), `method` ('multi' para inserção mais rápida com alguns dialetos).⁵³, ⁵⁴

````
# Exemplo conceitual: Salvar DataFrame no banco SQL com tipo de dado definido

# Importa o criador de engine (conexão) e tipos de dados específicos do SQLAlchemy
from sqlalchemy import create_engine, types

# Cria uma engine de conexão com banco SQLite local (um arquivinho .db simples)
engine_db = create_engine('sqlite:///meu_banco_escrita.db')

# Copia um DataFrame existente (para evitar alterar o original sem querer)
df_para_sql = df_pessoas.copy()

# Dicionário mapeando colunas do DataFrame para tipos SQL
dtype_map = {
    'Nome': types.VARCHAR(100),  # Coluna 'Nome' será texto com limite de 100 caracteres
    'Idade': types.INTEGER(),    # Coluna 'Idade' será um inteiro
    'Cidade': types.TEXT()       # Coluna 'Cidade' será texto livre (sem limite)
}

try:
    # Exporta o DataFrame para uma tabela SQL com controle de tipos
    df_para_sql.to_sql(
        'tabela_pessoas',       # Nome da tabela no banco de dados
        con=engine_db,          # Conexão do banco (a engine que criamos)
        if_exists='replace',    # Se a tabela já existir, apaga e recria (cuidado!)
        index=False,            # Não exporta o índice do DataFrame como coluna na tabela
        dtype=dtype_map,        # Aplica os tipos definidos para cada coluna
        method='multi'          # Envia várias linhas por vez (batch), melhora performance em alguns casos
    )
    print("\nDataFrame salvo em tabela SQL 'tabela_pessoas'.")

except Exception as e:
    # Captura qualquer erro que ocorrer durante o processo e imprime
    print(f"Erro ao salvar em SQL: {e}")

finally:
    # Encerra a conexão com o banco, liberando recursos
    if 'engine_db' in locals():
        engine_db.dispose()
````

### 🧠 Extras que vale saber:

- `method='multi'`: Nem todo banco suporta isso, mas quando suporta, é bala — ele insere **várias linhas em uma só query**. Bem mais rápido.
- `types.VARCHAR`, `types.INTEGER`: São classes da SQLAlchemy que definem como o SQL vai criar as colunas.
- `create_engine('sqlite:///...')`: Para SQLite local. Mas pode ser PostgreSQL, MySQL, etc. (ex: `'postgresql://user:senha@host:porta/db'`)

````
print("\nExemplos de Indexação Booleana:")

import pandas as pd

# Criando um DataFrame de exemplo
df = pd.DataFrame({
    'ID': range(1, 11),  # IDs de 1 a 10
    'Genero': ['F', 'M', 'F', 'M', 'F', 'F', 'M', 'F', 'M', 'F'],  # Gêneros variados
    'Pontuacao_Gasto': [75, 65, 85, 40, 90, 30, 60, 95, 70, 80],  # Pontuação de gasto
    'Renda_Anual_Mil': [95, 88, 120, 45, 100, 55, 70, 150, 85, 110],  # Renda em milhares
    'Categoria_Pref': [
        'Eletrônicos', 'Roupas', 'Livros', 'Alimentos',
        'Viagens', 'Eletrônicos', 'Livros', 'Viagens',
        'Roupas', 'Eletrônicos'  # Preferência de categoria
    ],
    'Idade': [25, 32, 29, 45, 37, 22, 34, 39, 28, 41]  # Idades diversas
})

# -------------------------------------------
# 1. Filtro complexo usando operadores lógicos
# -------------------------------------------

# Cria um DataFrame filtrado com base nas seguintes condições:
# - Gênero feminino ('F')
# - E (Pontuação de Gasto > 70 OU Renda Anual acima de 90 mil)
filtro_complexo = df[
    (df['Genero'] == 'F') &  # Verifica se a pessoa é do gênero feminino
    (
        (df['Pontuacao_Gasto'] > 70) |  # OU: se a pontuação de gasto for maior que 70
        (df['Renda_Anual_Mil'] > 90)    # OU: se a renda anual for maior que 90 mil
    )
]

# Mostra as colunas relevantes para esse filtro, usando .head() para exibir só os primeiros resultados
print("Feminino E (Score>70 OU Renda>90k):\n",
      filtro_complexo[['ID', 'Genero', 'Pontuacao_Gasto', 'Renda_Anual_Mil']].head())


# ---------------------------------------------------------
# 2. Filtro com .isin() para categorias específicas (OU)
# ---------------------------------------------------------

# Define uma lista com as categorias de interesse
categorias_interesse = ['Eletrônicos', 'Livros']

# Filtra o DataFrame para manter apenas as linhas onde 'Categoria_Pref'
# contém um dos valores da lista 'categorias_interesse'
filtro_isin = df[df['Categoria_Pref'].isin(categorias_interesse)]

# Exibe os primeiros resultados contendo apenas as colunas 'ID' e 'Categoria_Pref'
print(f"\nClientes com Categoria em {categorias_interesse}:\n",
      filtro_isin[['ID', 'Categoria_Pref']].head())


# ------------------------------------------
# 3. Filtro com .between() (intervalo inclusivo)
# ------------------------------------------

# Filtra os clientes cuja idade está entre 30 e 40 anos, inclusive
filtro_between = df[df['Idade'].between(30, 40)]  # between inclui os limites inferior e superior

# Exibe os primeiros clientes que atendem à condição, mostrando ID e idade
print("\nClientes com idade entre 30 e 40:\n",
      filtro_between[['ID', 'Idade']].head())
````

````
import pandas as pd
import numpy as np

print("\n--- 7. Manipulação de Dados (Adicional) ---")

# Simulando um DataFrame limpo
df_clean = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Idade': [25, 40, 30, 35, 28],
    'Genero': ['M', 'F', 'M', 'F', 'M'],
    'Categoria_Pref': ['Viagens', 'Eletrônicos', 'Livros', 'Vestuário', 'Vinhos'],
    'Pontuacao_Gasto': [60, 90, 70, 85, 75],
    'Renda_Anual_Mil': [70, 120, 80, 95, 65]
})

df_manip = df_clean.copy()

# --- .assign() ---
# Cria novas colunas de forma encadeada sem modificar o DataFrame original
# Sintaxe: df.assign(NOME_COLUNA=função_ou_valor)
# A função recebe o DataFrame atual como argumento (com lambda)
df_assign = df_manip.assign(
    # np.log1p(x) → log(1 + x), evita problema com log(0)
    Renda_Log = lambda x: np.log1p(x['Renda_Anual_Mil']),

    # Normaliza score para o intervalo [0,1]
    Score_Normalizado = lambda x: (
        (x['Pontuacao_Gasto'] - x['Pontuacao_Gasto'].min()) /
        (x['Pontuacao_Gasto'].max() - x['Pontuacao_Gasto'].min())
    ),

    # Multiplica Idade pela Pontuação → Feature Engineering puro
    Idade_x_Score = lambda x: x['Idade'] * x['Pontuacao_Gasto']
)

print("\nNovas colunas com .assign():")
print(df_assign[['Renda_Log', 'Score_Normalizado', 'Idade_x_Score']].head().round(2))


# --- .apply() ---
# Aplica uma função a cada coluna (axis=0) ou linha (axis=1)
# Muito usado para aplicar transformações em massa
# Sintaxe: df.apply(funcao, axis=0 ou 1)
# axis=0 → coluna por coluna (padrão)
# axis=1 → linha por linha (cada linha vira uma Series)

def log_transform(coluna):
    # Checa se a coluna é numérica (evita erro ao aplicar log em strings)
    if pd.api.types.is_numeric_dtype(coluna):
        return np.log1p(coluna)
    return coluna

df_log = df_clean.apply(log_transform, axis=0)
print("\nDataFrame após aplicar log1p a colunas numéricas (.apply axis=0):")
print(df_log.head().round(2))


# --- .applymap() ---
# Aplica uma função célula por célula (elemento a elemento)
# CUIDADO: só funciona com DataFrames (não com Series)
# Ideal para aplicar funções simples como limpeza de strings
# OBS: Está caindo em desuso — prefira .map() para Series e .apply() com axis

def para_maiuscula(valor):
    # Se for string, converte pra maiúscula
    if isinstance(valor, str):
        return valor.upper()
    return valor

df_upper = df_clean.applymap(para_maiuscula)
print("\nDataFrame após .applymap(para_maiuscula):")
print(df_upper.head())
````

**8. Grouping e Agregação (`.groupby()`)** (Expandido)

(Exemplos básicos de `.groupby().agg()` já na versão anterior, adicionamos `.transform()`, `.filter()`)

````
import pandas as pd
import numpy as np

print("\n--- 8. Grouping e Agregação (Adicional) ---")

# Criando um DataFrame de exemplo para funcionar com o código
df_final = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5, 6],
    'Regiao': ['Norte', 'Sul', 'Sul', 'Norte', 'Sudeste', 'Sudeste'],
    'Genero': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Idade': [25, 35, 45, 30, 40, 29],
    'Score': [50, 60, 70, 55, 80, 65],
    'Renda_k_Ano': [60, 80, 100, 70, 90, 85]
})

# Copiando para manter o original intacto
df_agg = df_final.copy()

# ===============================
# .transform(): Agrupa e retorna um resultado com o MESMO número de linhas do DataFrame original.
# Aqui, agrupamos por 'Regiao' e aplicamos a média ('mean') na coluna 'Renda_k_Ano'.
# O resultado será uma nova coluna onde cada linha recebe a média da sua própria região.
# Isso é útil quando queremos agregar mas manter o alinhamento com os dados originais.

df_agg['Renda_Media_Regiao'] = df_agg.groupby('Regiao')['Renda_k_Ano'].transform('mean')

print("\nDataFrame com Renda Média da Região (.transform()):")
print(df_agg[['Regiao', 'Renda_k_Ano', 'Renda_Media_Regiao']].head().round(1))


# ===============================
# .filter(): Serve para eliminar grupos inteiros com base em uma condição.
# Neste exemplo, usamos lambda para manter SOMENTE as regiões cuja média da coluna 'Score' > 55.
# A função é aplicada a cada grupo (grupo = subset do DataFrame com a mesma 'Regiao').
# Retorna um DataFrame somente com os grupos que passam no filtro.

df_filtrado_grupo = df_agg.groupby('Regiao').filter(lambda grupo: grupo['Score'].mean() > 55)

print("\nDataFrame filtrado por Grupo (.filter() - Regiões com Score Médio > 55):")
print("Regiões Mantidas:", df_filtrado_grupo['Regiao'].unique())
print(f"Shape Original: {df_agg.shape}, Shape Filtrado: {df_filtrado_grupo.shape}")


# ===============================
# .agg(): Usado com groupby para aplicar funções agregadas a cada grupo.
# Podemos passar:
# - Uma função simples ('mean', 'sum', etc)
# - Uma lista de funções ['mean', 'median']
# - Uma função customizada, como faremos agora

# Função customizada: intervalo interquartil (IQR)
def range_iqr(series):
    q1 = series.quantile(0.25)  # 1º quartil
    q3 = series.quantile(0.75)  # 3º quartil
    return q3 - q1              # Diferença = IQR

# Aplicamos a agregação agrupando por 'Genero' e usando:
# - média ('mean')
# - mediana ('median')
# - a função customizada range_iqr definida acima

print("\nAgregação customizada (IQR da Idade por Genero):")
print(df_agg.groupby('Genero')['Idade'].agg(['mean', 'median', range_iqr]))
````

**9. Combinação de DataFrames** (Expandido)

(Exemplos básicos de `concat`, `merge` já na versão anterior, adicionamos detalhes)

````
import pandas as pd

print("\n--- 9. Combinação de DataFrames (Adicional) ---")

# ======================================
# Criando dois DataFrames simples com colunas e índices diferentes
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']}, index=[0, 1])
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'C': ['C2', 'C3']}, index=[0, 2])  # Índice 2 e coluna 'C' diferente

# ======================================
# pd.concat(): concatena DataFrames ao longo do eixo 0 (linhas) ou 1 (colunas)
# Parâmetros principais:
# - objs: lista de DataFrames
# - axis: 0 para empilhar verticalmente (default), 1 para juntar lado a lado
# - join: 'outer' (default) une todas colunas, 'inner' só mantém colunas comuns
# - keys: adiciona um nível de índice (MultiIndex) identificando a origem

# Aqui, empilhamos df1 e df2 verticalmente com keys = ['fonte1', 'fonte2']
concat_keys = pd.concat([df1, df2], keys=['fonte1', 'fonte2'])

print("\nConcatenação Vertical com Keys (MultiIndex):\n", concat_keys)
# MultiIndex: o primeiro nível é 'fonte1' ou 'fonte2', o segundo é o índice original (0, 1, etc)

# Podemos acessar as "fatias" do MultiIndex com .loc
print("\nAcessando via MultiIndex (concat_keys.loc['fonte1']):\n", concat_keys.loc['fonte1'])

# ======================================
# Concatenação horizontal (lado a lado), unindo pelas linhas (índices)
# join='inner' mantém apenas os índices que estão em AMBOS os DataFrames

concat_h_inner = pd.concat([df1, df2], axis=1, join='inner')
print("\nConcatenação Horizontal (join='inner') - Apenas índices comuns:\n", concat_h_inner)
# Apenas o índice 0 é comum entre df1 e df2

# ======================================
# Merge entre dois DataFrames com múltiplas chaves ('key1' e 'key2')

df_left = pd.DataFrame({
    'key1': ['K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0'],
    'A': ['A0', 'A1', 'A2']
})

df_right = pd.DataFrame({
    'key1': ['K0', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0'],
    'B': ['B0', 'B1', 'B2']
})

print("\nDataFrames para Merge Múltiplo:\ndf_left:\n", df_left)
print("df_right:\n", df_right)

# pd.merge(): semelhante ao SQL JOIN
# Parâmetros:
# - on: colunas para o merge (pode ser uma lista para múltiplas)
# - how: tipo de join ('inner', 'left', 'right', 'outer')
# - indicator: True adiciona coluna '_merge' indicando a origem da linha
# - suffixes: sufixos para colunas com o mesmo nome nos dois DataFrames

merge_multi = pd.merge(
    df_left, df_right,
    on=['key1', 'key2'],  # merge por múltiplas colunas
    how='outer',          # mantém todas as combinações possíveis
    indicator=True,       # adiciona coluna '_merge'
    suffixes=('_esq', '_dir')  # sufixos se houver nomes duplicados de coluna
)
print("\nMerge Outer com Múltiplas Chaves e Indicator:\n", merge_multi)

# ======================================
# Merge entre índice de um DataFrame e coluna de outro

# Primeiro, definimos o índice de df_left como 'key1'
df_left_idx = df_left.set_index('key1')

# Copiamos df_right para manter original
df_right_col = df_right.copy()

# Aqui fazemos o merge entre:
# - o índice de df_left_idx
# - a coluna 'key1' de df_right_col
# Isso é útil quando temos DataFrames estruturados de forma diferente

merge_idx_col = pd.merge(
    df_left_idx, df_right_col,
    left_index=True,    # usa o índice de df_left_idx ('key1')
    right_on='key1',    # usa a coluna 'key1' de df_right_col
    how='left',         # todas as linhas da esquerda são mantidas
    suffixes=('_cli', '_comp')
)
print("\nMerge Índice com Coluna:\n", merge_idx_col)
````

**10. Tratamento de Dados Ausentes (Revisão/Expansão)**

(Exemplos básicos já vistos, adicionamos `limit`, `interpolate`)

````
import pandas as pd
import numpy as np

print("\n--- 10. Tratamento de Dados Ausentes (Adicional) ---")

# ===========================================
# Criando um DataFrame com valores ausentes (NaN)
# np.nan representa um valor "não numérico" (Not a Number)
df_miss = pd.DataFrame({
    'A': [1, 2, np.nan, np.nan, 5, np.nan],
    'B': [10, np.nan, 30, 40, np.nan, 60]
})
print("\nDataFrame para Dados Ausentes:\n", df_miss)

# ===========================================
# .fillna(): Preenche valores ausentes (NaN) com uma estratégia
# Parâmetros úteis:
# - method='ffill': "forward fill", copia o valor anterior não-nulo para frente
# - limit=N: só preenche até N NaNs consecutivos
# - value=algum_valor: define um valor fixo para preenchimento
# - inplace=True: altera o próprio DataFrame (caso contrário retorna um novo)

# Aqui fazemos forward fill com limit=1 → só preenche 1 NaN por vez
print("\nfillna(method='ffill', limit=1):\n", df_miss.fillna(method='ffill', limit=1))

# Resultado:
#   A: preenche o primeiro NaN com o valor acima (2), mas deixa o segundo
#   B: preenche o NaN da linha 1 com 10, mas pula os demais

# ===========================================
# .interpolate(): interpolação linear entre valores conhecidos
# Ideal para séries temporais ou dados numéricos contínuos
# Parâmetros:
# - method='linear' (default): interpolação simples entre dois pontos
# - limit: número máximo de valores consecutivos a preencher
# - axis: 0 (linha, default) ou 1 (coluna)
# - inplace=True: altera o DataFrame original
# ⚠️ Cuidado: se o primeiro ou último valor for NaN, ele pode não ser preenchido

# Aplicamos interpolação na coluna 'A'
df_miss['A_interpolado'] = df_miss['A'].interpolate(method='linear')

print("\nDataFrame com Coluna A Interpolada:\n", df_miss)

# Explicação da interpolação:
# A = [1.0, 2.0, NaN, NaN, 5.0, NaN]
# A interpolado fica:
# linha 2: 2.0 + (5.0 - 2.0) * 1/3 = 3.0
# linha 3: 2.0 + (5.0 - 2.0) * 2/3 = 4.0
# linha 5: continua NaN pois não há um valor depois de 5 para interpolar
````

🧠 Dicas ninja:

- `fillna()` é ótimo para preencher com médias, mediana, moda ou valores fixos. Exemplo:
  ```python
  df['coluna'].fillna(df['coluna'].mean())
  ```
- `interpolate()` é melhor para **tendência numérica contínua**, como séries temporais.
- Evite `fillna(method='ffill')` sem limite em dados com muitos NaNs no início — pode propagar valores irreais.
- Para apagar linhas ou colunas com muitos NaNs:
  ```python
  df.dropna(axis=0, how='any')   # remove linhas com pelo menos 1 NaN
  df.dropna(axis=1, how='all')   # remove colunas completamente vazias
  ```


````
import pandas as pd
import numpy as np

print("\n--- 10. Tratamento de Dados Ausentes (Avançado) ---")

# ===========================================
# Criando um DataFrame com valores ausentes (NaN)
df_miss = pd.DataFrame({
    'A': [1, 2, np.nan, np.nan, 5, np.nan],
    'B': [10, np.nan, 30, 40, np.nan, 60]
})
print("\nDataFrame para Dados Ausentes:\n", df_miss)

# ===========================================
# .dropna(): Remove linhas ou colunas com valores ausentes (NaN)
# Parâmetros:
# - axis=0 (linha, default) ou axis=1 (coluna)
# - how='any': remove linhas/colunas com pelo menos um NaN
# - how='all': remove linhas/colunas com TODOS os valores como NaN
# - thresh=N: mantém as linhas/colunas com pelo menos N valores não-NaN

# Exemplo: Remover linhas com qualquer NaN
df_drop_any = df_miss.dropna(axis=0, how='any')
print("\nRemovendo Linhas com Qualquer NaN (.dropna(axis=0, how='any')):\n", df_drop_any)

# Exemplo: Remover colunas com TODOS os valores NaN
df_drop_all = df_miss.dropna(axis=1, how='all')
print("\nRemovendo Colunas com Todos os Valores NaN (.dropna(axis=1, how='all')):\n", df_drop_all)

# Exemplo: Manter linhas com pelo menos 2 valores não-NaN
df_drop_thresh = df_miss.dropna(axis=0, thresh=2)
print("\nMantendo Linhas com Pelo Menos 2 Valores Não-NaN (.dropna(thresh=2)):\n", df_drop_thresh)

# ===========================================
# .isna() / .notna(): Identifica valores ausentes ou não ausentes
# .isna(): retorna True onde há NaN
# .notna(): retorna True onde NÃO há NaN

# Exemplo de uso de .isna() para identificar NaNs
isna_result = df_miss.isna()
print("\nResultado de .isna() (True para NaN, False para valores presentes):\n", isna_result)

# Exemplo de uso de .notna() para identificar valores presentes
notna_result = df_miss.notna()
print("\nResultado de .notna() (True para valores presentes, False para NaN):\n", notna_result)

# ===========================================
# .mask(): Substitui valores com base em uma condição
# .mask(condição): substitui os valores que atendem à condição

# Exemplo: Substituir NaNs em 'A' por 0
df_mask = df_miss.copy()
df_mask['A'] = df_mask['A'].mask(df_mask['A'].isna(), 0)
print("\nDataFrame após .mask() para substituir NaNs por 0:\n", df_mask)

# ===========================================
# Técnicas de Imputação Avançada

# Imputação com valores de outras colunas: aqui usamos o valor médio de 'A' para preencher os NaNs de 'B'
df_imputacao = df_miss.copy()
df_imputacao['B'] = df_imputacao['B'].fillna(df_imputacao['A'].mean())  # Imputa NaNs de 'B' com a média de 'A'
print("\nDataFrame após imputação da Coluna 'B' com a média de 'A':\n", df_imputacao)

# Imputação usando o método de interpolação para 'B' (aplica a interpolação em uma coluna)
df_miss['B_interpolado'] = df_miss['B'].interpolate(method='linear')
print("\nDataFrame após interpolação linear na Coluna 'B':\n", df_miss)

# Imputação usando KNN (K-Nearest Neighbors) da biblioteca 'sklearn' (exemplo avançado):
from sklearn.impute import KNNImputer

# Criando um imputer com 2 vizinhos mais próximos
imputer = KNNImputer(n_neighbors=2)

# Imputando os valores ausentes usando KNN (em todo o DataFrame)
df_knn_imputado = pd.DataFrame(imputer.fit_transform(df_miss), columns=df_miss.columns)
print("\nDataFrame após imputação usando KNN:\n", df_knn_imputado)
````

**11. Séries Temporais (Básico)** (Expandido)

(Exemplo básico já visto, adicionamos detalhes e `.shift()`, `.rolling()`)

````
print("\n--- 11. Séries Temporais (Adicional) ---")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criando um DataFrame simulado com dados de visitas, renda e pontuação
df_clean = pd.DataFrame({
    'ID': range(1, 31),  # 30 visitas fictícias
    'Data_Visita': pd.date_range(start='2024-07-01', periods=30, freq='D'),  # Datas de 1 a 30 de julho de 2024
    'Renda_Anual_Mil': np.random.randint(50, 200, size=30),  # Renda em milhares
    'Pontuacao_Gasto': np.random.randint(0, 100, size=30)  # Pontuação de gasto aleatória
})

# Cópia do DataFrame limpo (df_clean), onde assumimos que já foi tratado antes
df_ts = df_clean.copy()

# Converte a coluna 'Data_Visita' para o tipo datetime (fundamental para trabalhar com séries temporais!)
# Isso permite realizar filtros, ordenações e operações por tempo com eficiência
df_ts['Data_Visita'] = pd.to_datetime(df_ts['Data_Visita'])

# Define a coluna 'Data_Visita' como índice do DataFrame
# O Pandas trata índices do tipo datetime de forma especial, permitindo indexação e reamostragem por datas
df_ts.set_index('Data_Visita', inplace=True)

# Ordena o índice (ou seja, as datas) em ordem crescente
# Isso é obrigatório para que operações temporais funcionem corretamente (ex: rolling, shift, etc.)
df_ts.sort_index(inplace=True)

print("\nDataFrame Série Temporal (head):\n", df_ts.head())

# -----------------------------------------
# Seleção parcial por data
# Aqui usamos uma feature top do Pandas: seleção de datas com strings parciais.
# Por exemplo, '2024-07' pega tudo que for Julho de 2024.
print("\nCompras em 2024-Julho:")
print(df_ts.loc['2024-07'].head())  # ATENÇÃO: Isso só funciona se o índice for datetime

# -----------------------------------------
# Reamostragem semanal: agrupar os dados por semana (com início na segunda-feira)
# 'W-MON' significa "weekly (semanal), starting on Monday"
# .resample() permite agregar dados por intervalos temporais (diário, mensal, anual, etc)
resample_semanal = df_ts.resample('W-MON').agg(
    Contagem=('ID', 'size'),  # Conta o número de visitas (baseado no número de IDs por semana)
    Renda_Total_Semana=('Renda_Anual_Mil', 'sum')  # Soma o total de Renda Anual para cada semana
)
print("\nResample Semanal (Contagem e Soma Renda):")
print(resample_semanal.head())

# -----------------------------------------
# Deslocamento temporal com .shift()
# Muito útil em séries temporais para criar features como "valor anterior" (lag) ou "valor futuro" (lead)
# Neste exemplo, criamos uma nova coluna com o valor anterior da pontuação
df_ts['Score_Anterior'] = df_ts['Pontuacao_Gasto'].shift(1)

# .shift(1): move os valores 1 linha para baixo (lag de 1). A primeira linha vira NaN, pois não tem anterior.
print("\nDataFrame com Score da Visita Anterior (.shift(1)):")
print(df_ts[['Pontuacao_Gasto', 'Score_Anterior']].head())

# -----------------------------------------
# Janela móvel (Rolling Window): cálculo de estatísticas em janelas de tempo móveis
# Antes, precisamos garantir que temos **dados contínuos no tempo**, o que raramente acontece em dados brutos
# Por isso fazemos um .resample() diário, seguido de .ffill() para preencher os valores ausentes

# Reamostragem para diário (D = daily), preenchendo os dias faltantes com o valor anterior (.ffill)
df_ts_diario = df_ts['Renda_Anual_Mil'].resample('D').mean().ffill()

# Calcula média móvel de 7 dias:
# .rolling(window=7): cria janelas de 7 dias (pode ser baseada em observações ou tempo real)
# .mean(): aplica a média sobre cada janela
# min_periods=1: exige no mínimo 1 valor para calcular a média (senão seria NaN até preencher 7)
df_ts['Renda_Media_Movel_7D'] = df_ts_diario.rolling(window=7, min_periods=1).mean()

print("\nDataFrame com Média Móvel de 7 dias da Renda (.rolling(7).mean()):")
print(df_ts[['Renda_Anual_Mil', 'Renda_Media_Movel_7D']].dropna().head())
# dropna() usado aqui apenas para exibir onde realmente começa a série preenchida

# -----------------------------------------
# Visualização com matplotlib
# Import necessário (caso não tenha sido feito antes):
import matplotlib.pyplot as plt

# Plota a série original interpolada (diária) e a média móvel de 7 dias
plt.figure(figsize=(12, 5))
df_ts['Renda_Anual_Mil'].plot(label='Renda Original (Diária Interpolada)', alpha=0.5)
df_ts['Renda_Media_Movel_7D'].plot(label='Média Móvel 7 Dias', color='red')
plt.title('Renda Anual (Milhares) e Média Móvel 7 Dias')
plt.legend()
plt.show()
````

**13. Otimização de Performance**

Para DataFrames grandes, considere:

* **Tipos de Dados Eficientes:** Use dtypes menores quando possível (e.g., `int8`, `int16`, `float32`). Use `pd.Categorical` para strings com pouca variedade (economiza muita memória).

````
# === OTIMIZAÇÃO DE TIPOS DE DADOS (DTYPES) PARA ECONOMIZAR MEMÓRIA ===
import pandas as pd
import numpy as np

df_final = pd.DataFrame({
    'Idade': np.random.randint(18, 70, 1000),
    'Pontuacao_Gasto': np.random.randint(0, 100, 1000),
    'Anos_Cliente': np.random.randint(0, 20, 1000),
    'Genero': np.random.choice(['Feminino', 'Masculino'], 1000),
    'Regiao': np.random.choice(['Norte', 'Sul', 'Sudeste'], 1000),
    'Categoria_Pref': np.random.choice(['A', 'B', 'C'], 1000),
    'Faixa_Etaria': np.random.choice(['18-25', '26-35', '36-50', '50+'], 1000)
})

# Criamos uma cópia do DataFrame original para não modificar os dados brutos durante a otimização.
df_opt = df_final.copy()

# Exibe os tipos de dados originais de cada coluna.
print("\nDtypes Originais:")
print(df_opt.dtypes)

# Calcula o uso total de memória do DataFrame em megabytes.
# memory_usage(deep=True) faz uma análise mais precisa, incluindo strings e objetos.
# .sum() soma o uso de todas as colunas.
# Dividimos por 1024^2 para converter bytes em megabytes.
print(f"Uso de memória inicial: {df_opt.memory_usage(deep=True).sum() / 1024**2:.2f} MB")


# === CONVERSÃO DE INTEIROS PARA TIPOS MENORES ===

# Aqui, reduzimos o tamanho das colunas numéricas inteiras.
# Se os valores cabem dentro de um intervalo menor, podemos trocar de tipo.
# Isso economiza bastante memória, especialmente em grandes datasets.

# np.int8 = número inteiro entre -128 e 127 (1 byte)
# Obs: Se a coluna tiver valores maiores que 127 ou menores que -128, vai dar erro.

df_opt['Idade'] = df_opt['Idade'].astype(np.int8)  # exemplo: se Idade varia entre 0 e 100
df_opt['Pontuacao_Gasto'] = df_opt['Pontuacao_Gasto'].astype(np.int8)  # idem para Score/Pontuação
df_opt['Anos_Cliente'] = df_opt['Anos_Cliente'].astype(np.int8)  # idem para tempo de cliente

# Alternativas:
# np.int16: -32.768 a 32.767 (2 bytes)
# np.int32: -2 bi a 2 bi (4 bytes)
# np.int64: padrão do Pandas (8 bytes), só use se necessário


# === CONVERSÃO DE STRINGS/CATEGORIAS PARA TYPE 'category' ===

# Colunas de texto com **baixa cardinalidade** (poucos valores únicos) são candidatas ideais para o tipo 'category'.
# Isso reduz drasticamente o uso de memória e acelera operações como groupby, merge e filtragem.
# Exemplo: Gênero = ['Masculino', 'Feminino', 'Outro'] → ideal p/ category

for col in ['Genero', 'Regiao', 'Categoria_Pref', 'Faixa_Etaria']:
    if col in df_opt.columns:  # verifica se a coluna existe no DataFrame
        df_opt[col] = df_opt[col].astype('category')  # converte para tipo categórico

# Alternativas:
# - 'category' com parâmetro ordered=True para ordens específicas (ex: Baixo < Médio < Alto)
# - Cuidado ao usar com muitas categorias únicas (alta cardinalidade), pois pode piorar a performance


# === RESULTADOS APÓS A OTIMIZAÇÃO ===

print("\nDtypes Otimizados:")
print(df_opt.dtypes)

# Recalcula e imprime o novo uso de memória após otimizar os tipos.
print(f"Uso de memória otimizado: {df_opt.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
````

* **Operações Vetorizadas:** Evite iterar sobre linhas (`.iterrows()`, `.apply(axis=1)` é lento). Use operações NumPy/Pandas que atuam no array/Series inteiro.
* **Indexação:** Use indexação eficiente (`.loc`, `.iloc`, booleana). Índices ordenados (`.sort_index()`) podem acelerar seleções baseadas em rótulos.
* **Bibliotecas Externas:** Para dados maiores que a memória, considere Dask, Vaex ou processamento distribuído (Spark com PySpark). Para acelerar código Python numérico, considere Numba ou Cython.
